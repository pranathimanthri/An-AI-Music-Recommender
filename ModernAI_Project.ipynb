```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, Flatten, Dense, Dropout, concatenate
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt

# Load the Spotify data
spotify_data = pd.read_csv('spotify_data.csv')
print("Features:", spotify_data.columns)

# Basic data exploration
print(f"Number of tracks: {len(spotify_data)}")
print(f"Number of unique artists: {len(spotify_data['artist_name'].unique())}")

# Preprocess data
# Select relevant columns (adjust based on your actual data)
feature_columns = [
    'danceability', 'energy', 'loudness', 'speechiness',
    'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo'
]

# Check for missing values and drop them
spotify_data = spotify_data.dropna(subset=feature_columns + ['track_name', 'artist_name'])

# Create user-item matrix (for collaborative filtering)
# In a real scenario, you'd have user-track interactions, but for this demo we'll create a dummy user ID
# This is just for demonstration - in practice you need real user-track interaction data
np.random.seed(42)
n_users = 1000
spotify_data['user_id'] = np.random.randint(1, n_users+1, size=len(spotify_data))

# Convert categorical data to numerical IDs
from sklearn.preprocessing import LabelEncoder

user_encoder = LabelEncoder()
spotify_data['user_id'] = user_encoder.fit_transform(spotify_data['user_id'])
num_users = spotify_data['user_id'].nunique()

item_encoder = LabelEncoder()
spotify_data['track_id_encoded'] = item_encoder.fit_transform(spotify_data['track_id'])
num_tracks = spotify_data['track_id_encoded'].nunique()

# Create a synthetic rating column (1-5 scale) for this example
# In practice, this would be actual user ratings or play counts
spotify_data['rating'] = np.random.uniform(1, 5, size=len(spotify_data))

# Split data into training and test sets
train_data, test_data = train_test_split(
    spotify_data[['user_id', 'track_id_encoded', 'rating']],
    test_size=0.2,
    random_state=42
)

# Prepare input arrays for neural network
X_train = [train_data['user_id'].values, train_data['track_id_encoded'].values]
y_train = train_data['rating'].values

X_test = [test_data['user_id'].values, test_data['track_id_encoded'].values]
y_test = test_data['rating'].values

# Build neural collaborative filtering model
# 1. Define model architecture
num_factors = 50  # Embedding size

# Input layers
user_input = Input(shape=(1,), name='user-input')
track_input = Input(shape=(1,), name='track-input')

# Embedding layers
user_embedding = Embedding(num_users, num_factors, name='user-embedding')(user_input)
user_vec = Flatten(name='user-flatten')(user_embedding)

track_embedding = Embedding(num_tracks, num_factors, name='track-embedding')(track_input)
track_vec = Flatten(name='track-flatten')(track_embedding)

# Concatenate embeddings
concat = concatenate([user_vec, track_vec])
concat_dropout = Dropout(0.2)(concat)

# Dense layers
fc_1 = Dense(100, activation='relu')(concat_dropout)
fc_1_dropout = Dropout(0.2)(fc_1)
fc_2 = Dense(50, activation='relu')(fc_1_dropout)
fc_2_dropout = Dropout(0.2)(fc_2)

# Output layer
output = Dense(1)(fc_2_dropout)

# Create and compile model
model = Model([user_input, track_input], output)
model.compile(loss='mean_squared_error', optimizer=Adam(learning_rate=0.001))

# Print model summary
model.summary()

# Train the model
history = model.fit(
    X_train, y_train,
    batch_size=64,
    epochs=5,
    verbose=1,
    validation_split=0.1
)

# Evaluate the model
loss = model.evaluate(X_test, y_test)
print(f"Test MSE: {loss}")
print(f"Test RMSE: {np.sqrt(loss)}")

# Plot training history
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.grid(True)
plt.show()

# Function to get track recommendations
def get_recommendations(user_id, n=10):
    """Get top n track recommendations for a user"""
    # Create input for all tracks for this user
    user_ids = np.full(num_tracks, user_id)
    track_ids = np.arange(num_tracks)

    # Predict ratings
    predictions = model.predict([user_ids, track_ids], verbose=0).flatten()

    # Get top n track indices
    top_indices = predictions.argsort()[-n:][::-1]

    # Get track information
    recommendations = []
    for idx in top_indices:
        track_info = spotify_data[spotify_data['track_id_encoded'] == idx].iloc[0]
        recommendations.append({
            'artist': track_info['artist_name'],
            'track': track_info['track_name'],
            'predicted_rating': predictions[idx]
        })

    return recommendations

# Example: Get recommendations for a random user
random_user = np.random.randint(0, num_users)
recommendations = get_recommendations(random_user)
print(f"\nTop recommendations for user {random_user}:")
for i, rec in enumerate(recommendations):
    print(f"{i+1}. {rec['artist']} - {rec['track']} (Score: {rec['predicted_rating']:.2f})")

# Content-based recommendation function using audio features
def get_similar_tracks(track_name, artist_name, n=10):
    """Find similar tracks based on audio features"""
    # Normalize features
    scaler = MinMaxScaler()
    normalized_features = scaler.fit_transform(spotify_data[feature_columns])

    # Find the seed track
    seed_track = spotify_data[
        (spotify_data['artist_name'].str.lower() == artist_name.lower()) &
        (spotify_data['track_name'].str.lower() == track_name.lower())
    ]

    if seed_track.empty:
        print(f"Track '{track_name}' by '{artist_name}' not found")
        return None

    # Get the index of the seed track
    seed_idx = seed_track.index[0]

    # Calculate cosine similarity
    from sklearn.metrics.pairwise import cosine_similarity
    seed_features = normalized_features[seed_idx].reshape(1, -1)
    similarities = cosine_similarity(seed_features, normalized_features)[0]

    # Get indices of top similar tracks (excluding the seed track)
    similar_indices = similarities.argsort()[-n-1:][::-1]
    similar_indices = [idx for idx in similar_indices if idx != seed_idx][:n]

    # Get track information
    similar_tracks = []
    for idx in similar_indices:
        track_info = spotify_data.iloc[idx]
        similar_tracks.append({
            'artist': track_info['artist_name'],
            'track': track_info['track_name'],
            'similarity': similarities[idx]
        })

    return similar_tracks

# Example: Find similar tracks
example_track = "Blinding Lights"
example_artist = "The Weeknd"
similar_tracks = get_similar_tracks(example_track, example_artist)

if similar_tracks:
    print(f"\nTracks similar to {example_artist} - {example_track}:")
    for i, track in enumerate(similar_tracks):
        print(f"{i+1}. {track['artist']} - {track['track']} (Similarity: {track['similarity']:.4f})")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import Input, Embedding, Flatten, Dense, Dropout, concatenate
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from sklearn.metrics.pairwise import cosine_similarity
import os

class SpotifyRecommenderFineTuner:
    def __init__(self, data_path='spotify_data.csv'):
        """Initialize the fine-tuning model"""
        self.data_path = data_path
        self.model = None
        self.spotify_data = None
        self.feature_columns = [
            'danceability', 'energy', 'loudness', 'speechiness',
            'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo'
        ]
        self.user_encoder = LabelEncoder()
        self.track_encoder = LabelEncoder()
        self.scaler = MinMaxScaler()

    def load_data(self):
        """Load and process Spotify data"""
        try:
            # Load data
            self.spotify_data = pd.read_csv(self.data_path)
            print(f"Successfully loaded {len(self.spotify_data)} tracks")
            print("Available columns:", list(self.spotify_data.columns))
            return True
        except Exception as e:
            print(f"Error loading data: {e}")
            return False

    def preprocess_data(self, create_synthetic_users=True, n_users=1000):
        """Preprocess data for model training"""
        if self.spotify_data is None:
            print("No data loaded. Call load_data() first.")
            return False

        # Handle missing values
        self.spotify_data = self.spotify_data.dropna(subset=self.feature_columns + ['track_name', 'artist_name'])

        # Normalize audio features
        self.spotify_data[self.feature_columns] = self.scaler.fit_transform(
            self.spotify_data[self.feature_columns]
        )

        # If we need to create synthetic user data for training
        if create_synthetic_users:
            np.random.seed(42)
            self.spotify_data['user_id'] = np.random.randint(1, n_users+1, size=len(self.spotify_data))
            # Create synthetic ratings (1-5 scale)
            self.spotify_data['rating'] = np.random.uniform(1, 5, size=len(self.spotify_data))

        # Encode categorical variables
        self.spotify_data['user_id'] = self.user_encoder.fit_transform(self.spotify_data['user_id'])
        self.spotify_data['track_id_encoded'] = self.track_encoder.fit_transform(self.spotify_data['track_id'])

        self.num_users = self.spotify_data['user_id'].nunique()
        self.num_tracks = self.spotify_data['track_id_encoded'].nunique()

        print(f"Preprocessed data with {len(self.spotify_data)} tracks, {self.num_users} users")
        return True

    def create_model(self, num_factors=50, learning_rate=0.001):
        """Create a neural collaborative filtering model"""
        # Input layers
        user_input = Input(shape=(1,), name='user-input')
        track_input = Input(shape=(1,), name='track-input')

        # Embedding layers
        user_embedding = Embedding(self.num_users, num_factors, name='user-embedding')(user_input)
        user_vec = Flatten(name='user-flatten')(user_embedding)

        track_embedding = Embedding(self.num_tracks, num_factors, name='track-embedding')(track_input)
        track_vec = Flatten(name='track-flatten')(track_embedding)

        # Concatenate embeddings
        concat = concatenate([user_vec, track_vec])
        concat_dropout = Dropout(0.2)(concat)

        # Dense layers
        fc_1 = Dense(100, activation='relu')(concat_dropout)
        fc_1_dropout = Dropout(0.2)(fc_1)
        fc_2 = Dense(50, activation='relu')(fc_1_dropout)
        fc_2_dropout = Dropout(0.2)(fc_2)

        # Output layer
        output = Dense(1)(fc_2_dropout)

        # Create and compile model
        self.model = Model([user_input, track_input], output)
        self.model.compile(loss='mean_squared_error', optimizer=Adam(learning_rate=learning_rate))

        print("Model created successfully")
        return self.model

    def train_model(self, epochs=10, batch_size=64, validation_split=0.1):
        """Train the model on the preprocessed data"""
        if self.model is None:
            print("No model created. Call create_model() first.")
            return None

        # Split data into training and validation sets
        train_data, test_data = train_test_split(
            self.spotify_data[['user_id', 'track_id_encoded', 'rating']],
            test_size=0.2,
            random_state=42
        )

        # Prepare input arrays
        X_train = [train_data['user_id'].values, train_data['track_id_encoded'].values]
        y_train = train_data['rating'].values

        X_test = [test_data['user_id'].values, test_data['track_id_encoded'].values]
        y_test = test_data['rating'].values

        # Set up callbacks for better training
        callbacks = [
            EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
            ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=0.00001),
            ModelCheckpoint('spotify_recommender_model.h5', monitor='val_loss', save_best_only=True)
        ]

        # Train the model
        history = self.model.fit(
            X_train, y_train,
            batch_size=batch_size,
            epochs=epochs,
            verbose=1,
            validation_split=validation_split,
            callbacks=callbacks
        )

        # Evaluate the model
        test_loss = self.model.evaluate(X_test, y_test)
        print(f"Test MSE: {test_loss}")
        print(f"Test RMSE: {np.sqrt(test_loss)}")

        # Plot training history
        plt.figure(figsize=(10, 6))
        plt.plot(history.history['loss'], label='Training Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.title('Model Loss During Training')
        plt.ylabel('Loss')
        plt.xlabel('Epoch')
        plt.legend()
        plt.grid(True)
        plt.show()

        self.X_test = X_test
        self.y_test = y_test

        return history

    def fine_tune(self, user_feedback, lr_multiplier=0.5, epochs=5):
        """
        Fine-tune the model based on explicit user feedback

        Parameters:
        - user_feedback: DataFrame with columns 'user_id', 'track_id', 'actual_rating'
        - lr_multiplier: Factor to reduce learning rate for fine-tuning
        - epochs: Number of fine-tuning epochs
        """
        if self.model is None:
            print("No trained model available. Train a model first.")
            return None

        # Prepare user feedback data
        user_feedback['user_id'] = self.user_encoder.transform(user_feedback['user_id'])
        user_feedback['track_id_encoded'] = self.track_encoder.transform(user_feedback['track_id'])

        X_feedback = [user_feedback['user_id'].values, user_feedback['track_id_encoded'].values]
        y_feedback = user_feedback['actual_rating'].values

        # Reduce learning rate for fine-tuning
        # In newer TensorFlow versions, we can't directly access lr this way
        # Instead, we'll just use the lr_multiplier to create a new optimizer
        new_lr = 0.001 * lr_multiplier  # Default Adam learning rate * multiplier
        self.model.compile(loss='mean_squared_error', optimizer=Adam(learning_rate=new_lr))

        # Fine-tune the model
        print(f"Fine-tuning model with {len(user_feedback)} feedback samples (learning rate: {new_lr})")
        history = self.model.fit(
            X_feedback, y_feedback,
            batch_size=32,
            epochs=epochs,
            verbose=1
        )

        # Re-evaluate on test set
        test_loss = self.model.evaluate(self.X_test, self.y_test)
        print(f"Post-fine-tuning Test MSE: {test_loss}")
        print(f"Post-fine-tuning Test RMSE: {np.sqrt(test_loss)}")

        return history

    def get_recommendations(self, user_id, n=10):
        """Get top n track recommendations for a user"""
        if self.model is None:
            print("No trained model available.")
            return None

        # Encode user ID
        if isinstance(user_id, (str, int)) and user_id not in self.user_encoder.classes_:
            print(f"User ID {user_id} not found in training data.")
            return None

        if isinstance(user_id, (str, int)):
            user_id = self.user_encoder.transform([user_id])[0]

        # Create input for all tracks for this user
        user_ids = np.full(self.num_tracks, user_id)
        track_ids = np.arange(self.num_tracks)

        # Predict ratings
        predictions = self.model.predict([user_ids, track_ids], verbose=0).flatten()

        # Get top n track indices
        top_indices = predictions.argsort()[-n:][::-1]

        # Get track information
        recommendations = []
        for idx in top_indices:
            # Get original track ID
            original_track_id = self.track_encoder.inverse_transform([idx])[0]

            # Find track info
            track_info = self.spotify_data[self.spotify_data['track_id'] == original_track_id].iloc[0]

            recommendations.append({
                'artist': track_info['artist_name'],
                'track': track_info['track_name'],
                'predicted_rating': predictions[idx],
                'track_id': original_track_id
            })

        return recommendations

    def get_similar_tracks(self, track_name, artist_name=None, track_id=None, n=10):
        """Find similar tracks based on audio features"""
        if self.spotify_data is None:
            print("No data loaded.")
            return None

        # Find the seed track
        if track_id is not None:
            seed_track = self.spotify_data[self.spotify_data['track_id'] == track_id]
        elif artist_name is not None:
            seed_track = self.spotify_data[
                (self.spotify_data['artist_name'].str.lower() == artist_name.lower()) &
                (self.spotify_data['track_name'].str.lower() == track_name.lower())
            ]
        else:
            seed_track = self.spotify_data[self.spotify_data['track_name'].str.lower() == track_name.lower()]

        if seed_track.empty:
            print(f"Track '{track_name}' not found in dataset")
            return None

        # Get the index of the seed track
        seed_idx = seed_track.index[0]

        # Extract and normalize features for similarity calculation
        track_features = self.spotify_data[self.feature_columns].values

        # Calculate cosine similarity
        seed_features = track_features[seed_idx].reshape(1, -1)
        similarities = cosine_similarity(seed_features, track_features)[0]

        # Get indices of top similar tracks (excluding the seed track)
        similar_indices = similarities.argsort()[-n-1:][::-1]
        similar_indices = [idx for idx in similar_indices if idx != seed_idx][:n]

        # Get track information
        similar_tracks = []
        for idx in similar_indices:
            track_info = self.spotify_data.iloc[idx]
            similar_tracks.append({
                'artist': track_info['artist_name'],
                'track': track_info['track_name'],
                'similarity': similarities[idx],
                'track_id': track_info['track_id']
            })

        return similar_tracks

    def save_model(self, path='spotify_recommender_fine_tuned.h5'):
        """Save the trained/fine-tuned model"""
        if self.model is None:
            print("No model to save.")
            return False

        try:
            self.model.save(path)
            print(f"Model saved to {path}")
            return True
        except Exception as e:
            print(f"Error saving model: {e}")
            return False

    def load_saved_model(self, path='spotify_recommender_fine_tuned.h5'):
        """Load a previously saved model"""
        try:
            self.model = load_model(path)
            print(f"Model loaded from {path}")
            return True
        except Exception as e:
            print(f"Error loading model: {e}")
            return False

    def explain_recommendation(self, recommendation):
        """Explain why a track was recommended"""
        if self.spotify_data is None:
            print("No data loaded.")
            return None

        track_id = recommendation['track_id']
        track_info = self.spotify_data[self.spotify_data['track_id'] == track_id].iloc[0]

        # Get feature values
        feature_values = {col: track_info[col] for col in self.feature_columns}

        # Determine the top features
        sorted_features = sorted(feature_values.items(), key=lambda x: x[1], reverse=True)
        top_features = sorted_features[:3]

        explanation = {
            'track': track_info['track_name'],
            'artist': track_info['artist_name'],
            'top_features': top_features,
            'similar_tracks': self.get_similar_tracks(track_name=None, track_id=track_id, n=3)
        }

        return explanation


# Example usage of the fine-tuning model
if __name__ == "__main__":
    # Initialize and train the model
    recommender = SpotifyRecommenderFineTuner('spotify_data.csv')
    recommender.load_data()
    recommender.preprocess_data()
    recommender.create_model()
    recommender.train_model(epochs=10)

    # Get recommendations for a user
    user_id = 1  # Example user ID
    recommendations = recommender.get_recommendations(user_id, n=5)
    print(f"\nTop 5 recommendations for user {user_id}:")
    for i, rec in enumerate(recommendations):
        print(f"{i+1}. {rec['artist']} - {rec['track']} (Predicted rating: {rec['predicted_rating']:.2f})")

    # Create sample user feedback for fine-tuning
    # This would normally come from real user interactions
    user_feedback = pd.DataFrame([
        {'user_id': user_id, 'track_id': recommendations[0]['track_id'], 'actual_rating': 2.0},  # User didn't like the top recommendation
        {'user_id': user_id, 'track_id': recommendations[1]['track_id'], 'actual_rating': 4.5},  # User really liked the second recommendation
        {'user_id': user_id, 'track_id': recommendations[2]['track_id'], 'actual_rating': 3.0},  # User was neutral about the third recommendation
    ])

    # Fine-tune the model with user feedback
    recommender.fine_tune(user_feedback, epochs=3)

    # Get updated recommendations
    updated_recommendations = recommender.get_recommendations(user_id, n=5)
    print(f"\nUpdated recommendations for user {user_id} after fine-tuning:")
    for i, rec in enumerate(updated_recommendations):
        print(f"{i+1}. {rec['artist']} - {rec['track']} (Predicted rating: {rec['predicted_rating']:.2f})")

    # Explain a recommendation
    explanation = recommender.explain_recommendation(updated_recommendations[0])
    print(f"\nWhy we recommended {explanation['artist']} - {explanation['track']}:")
    print(f"This track has strong {', '.join([f[0] for f in explanation['top_features']])}")
    print("Similar tracks you might enjoy:")
    for i, track in enumerate(explanation['similar_tracks']):
        print(f"- {track['artist']} - {track['track']}")

    # Save the fine-tuned model
    recommender.save_model()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, Flatten, Dense, Dropout, concatenate
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import seaborn as sns
import datetime

class SpotifyHistoryBasedRecommender:
    def __init__(self, history_path='history.csv', spotify_data_path='spotify_data.csv'):
        """Initialize the recommender using listening history data"""
        self.history_path = history_path
        self.spotify_data_path = spotify_data_path
        self.model = None
        self.history_data = None
        self.spotify_data = None
        self.feature_columns = [
            'danceability', 'energy', 'loudness', 'speechiness',
            'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo'
        ]
        self.track_encoder = LabelEncoder()
        self.artist_encoder = LabelEncoder()
        self.scaler = MinMaxScaler()

    def load_data(self):
        """Load listening history and Spotify audio features data"""
        try:
            # Load listening history
            self.history_data = pd.read_csv(self.history_path)
            print(f"Successfully loaded {len(self.history_data)} history records")

            # Load Spotify data with audio features
            self.spotify_data = pd.read_csv(self.spotify_data_path)
            print(f"Successfully loaded {len(self.spotify_data)} tracks with audio features")

            return True
        except Exception as e:
            print(f"Error loading data: {e}")
            return False

    def preprocess_history_data(self):
        """Preprocess listening history data"""
        if self.history_data is None:
            print("No history data loaded. Call load_data() first.")
            return False

        # Convert endTime to datetime
        self.history_data['endTime'] = pd.to_datetime(self.history_data['endTime'])

        # Clean track and artist names
        self.history_data['trackName'] = self.history_data['trackName'].str.strip()
        self.history_data['artistName'] = self.history_data['artistName'].str.strip()

        # Create a user_id column (since we only have one user's data)
        self.history_data['user_id'] = 1

        # Create an implicit rating based on play count or play time
        # Group by artist and track to count plays and total time
        play_stats = self.history_data.groupby(['artistName', 'trackName']).agg(
            play_count=('endTime', 'count'),
            total_time=('mPlayed', 'sum')
        ).reset_index()

        # Normalize play_count and total_time to create implicit ratings
        play_stats['play_count_norm'] = play_stats['play_count'] / play_stats['play_count'].max()
        play_stats['time_norm'] = play_stats['total_time'] / play_stats['total_time'].max()

        # Create an implicit rating (0.3*play_count + 0.7*time_spent)
        play_stats['implicit_rating'] = 0.3 * play_stats['play_count_norm'] + 0.7 * play_stats['time_norm']

        # Scale to 1-5 rating scale
        play_stats['rating'] = 1 + 4 * play_stats['implicit_rating']

        # Keep only the necessary columns
        self.user_history = play_stats[['artistName', 'trackName', 'rating', 'play_count', 'total_time']]

        print(f"Processed {len(self.user_history)} unique tracks in listening history")
        return True

    def merge_with_spotify_data(self):
        """Merge listening history with Spotify audio features"""
        if self.user_history is None or self.spotify_data is None:
            print("Both history and Spotify data need to be loaded and processed first.")
            return False

        # Merge on artist and track name
        self.merged_data = pd.merge(
            self.user_history,
            self.spotify_data,
            left_on=['artistName', 'trackName'],
            right_on=['artist_name', 'track_name'],
            how='inner'
        )

        match_rate = len(self.merged_data) / len(self.user_history) * 100
        print(f"Matched {len(self.merged_data)} tracks out of {len(self.user_history)} ({match_rate:.2f}%)")

        # For tracks that didn't match, try a more flexible approach (not implemented here)
        # For a production system, you would use fuzzy matching or additional identifiers

        if len(self.merged_data) == 0:
            print("No matches found between history and Spotify data. Check data formats.")
            return False

        return True

    def analyze_listening_patterns(self):
        """Analyze listening patterns to understand preferences"""
        if self.merged_data is None:
            print("No merged data available. Call merge_with_spotify_data() first.")
            return False

        # Create visualizations of listening patterns
        plt.figure(figsize=(12, 6))

        # Top artists by play count
        top_artists = self.history_data.groupby('artistName').size().sort_values(ascending=False).head(10)
        plt.subplot(1, 2, 1)
        top_artists.plot(kind='barh')
        plt.title('Top 10 Artists by Play Count')
        plt.xlabel('Play Count')
        plt.tight_layout()

        # Audio feature preferences
        plt.subplot(1, 2, 2)
        feature_averages = self.merged_data[self.feature_columns].mean()
        feature_averages.plot(kind='bar')
        plt.title('Average Audio Feature Values for Liked Tracks')
        plt.xticks(rotation=45)
        plt.tight_layout()

        plt.show()

        # Compute correlations between audio features and ratings
        corr = self.merged_data[self.feature_columns + ['rating']].corr()['rating'].drop('rating')

        print("\nCorrelation between audio features and ratings:")
        for feature, correlation in corr.sort_values(ascending=False).items():
            print(f"{feature}: {correlation:.4f}")

        # Most influential audio features
        print("\nMost influential audio features in your listening:")
        influential_features = corr.abs().sort_values(ascending=False).head(3).index.tolist()
        print(f"1. {influential_features[0]}")
        print(f"2. {influential_features[1]}")
        print(f"3. {influential_features[2]}")

        return influential_features

    def prepare_train_test_data(self):
        """Prepare training and test data for the model"""
        if self.merged_data is None:
            print("No merged data available. Call merge_with_spotify_data() first.")
            return False

        # Add user_id column
        self.merged_data['user_id'] = 1

        # Encode categorical variables
        self.merged_data['track_id_encoded'] = self.track_encoder.fit_transform(self.merged_data['track_id'])
        self.num_tracks = self.merged_data['track_id_encoded'].nunique()

        # Split data into training and test sets
        train_data, test_data = train_test_split(
            self.merged_data[['user_id', 'track_id_encoded', 'rating']],
            test_size=0.2,
            random_state=42
        )

        # Prepare input arrays
        self.X_train = [train_data['user_id'].values, train_data['track_id_encoded'].values]
        self.y_train = train_data['rating'].values

        self.X_test = [test_data['user_id'].values, test_data['track_id_encoded'].values]
        self.y_test = test_data['rating'].values

        print(f"Prepared {len(train_data)} training samples and {len(test_data)} test samples")
        return True

    def build_model(self, num_factors=50):
        """Build a collaborative filtering model based on one user's listening history"""
        # We'll use a simplified model since we only have one user
        # In this case, the model learns track embeddings based on the audio features

        # Input layers
        track_input = Input(shape=(1,), name='track-input')

        # Embedding layer for tracks
        track_embedding = Embedding(self.num_tracks, num_factors, name='track-embedding')(track_input)
        track_vec = Flatten(name='track-flatten')(track_embedding)

        # Dense layers
        dense_1 = Dense(64, activation='relu')(track_vec)
        dropout_1 = Dropout(0.2)(dense_1)
        dense_2 = Dense(32, activation='relu')(dropout_1)
        dropout_2 = Dropout(0.2)(dense_2)

        # Output layer
        output = Dense(1)(dropout_2)

        # Create and compile model
        # For a single user, we use a simplified model structure
        self.model = Model(track_input, output)
        self.model.compile(loss='mean_squared_error', optimizer=Adam(learning_rate=0.001))

        print("Model built successfully:")
        self.model.summary()
        return True

    def train_model(self, epochs=20, batch_size=32, validation_split=0.1):
        """Train the model on the preprocessed data"""
        if self.model is None:
            print("No model created. Call build_model() first.")
            return None

        # Set up callbacks for better training
        callbacks = [
            EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
            ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=0.00001)
        ]

        # Train the model
        # We need to adapt the training for a single-user case
        history = self.model.fit(
            self.X_train[1],  # Only use track IDs since we have a single user
            self.y_train,
            batch_size=batch_size,
            epochs=epochs,
            verbose=1,
            validation_split=validation_split,
            callbacks=callbacks
        )

        # Evaluate the model
        test_loss = self.model.evaluate(self.X_test[1], self.y_test)
        print(f"Test MSE: {test_loss}")
        print(f"Test RMSE: {np.sqrt(test_loss)}")

        # Plot training history
        plt.figure(figsize=(10, 6))
        plt.plot(history.history['loss'], label='Training Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.title('Model Loss During Training')
        plt.ylabel('Loss')
        plt.xlabel('Epoch')
        plt.legend()
        plt.grid(True)
        plt.show()

        return history

    def recommend_new_tracks(self, n=10):
        """Recommend new tracks based on listening history"""
        if self.model is None:
            print("No trained model available.")
            return None

        # Find tracks in Spotify data that are not in the user's history
        history_track_ids = set(self.merged_data['track_id'].unique())
        new_tracks = self.spotify_data[~self.spotify_data['track_id'].isin(history_track_ids)]

        if len(new_tracks) == 0:
            print("No new tracks found in Spotify data.")
            return None

        # Encode track IDs
        try:
            new_tracks['track_id_encoded'] = self.track_encoder.transform(new_tracks['track_id'])
        except ValueError:
            # Handle tracks with IDs not seen during training
            print("Some track IDs were not seen during training. Using audio feature similarity instead.")
            return self.recommend_by_audio_features(n)

        # Predict ratings for new tracks
        track_ids = new_tracks['track_id_encoded'].values
        predictions = self.model.predict(track_ids, verbose=0).flatten()

        # Get top n track indices
        top_indices = predictions.argsort()[-n:][::-1]

        # Get track information
        recommendations = []
        for i, idx in enumerate(top_indices):
            track_info = new_tracks.iloc[idx]
            recommendations.append({
                'rank': i + 1,
                'artist': track_info['artist_name'],
                'track': track_info['track_name'],
                'predicted_rating': predictions[idx],
                'track_id': track_info['track_id']
            })

        return recommendations

    def recommend_by_audio_features(self, n=10):
        """Recommend tracks based on audio feature similarity"""
        if self.merged_data is None or self.spotify_data is None:
            print("Both merged data and Spotify data need to be loaded first.")
            return None

        # Calculate user's preferred audio feature profile (weighted by ratings)
        user_profile = self.merged_data.copy()

        # Normalize ratings to use as weights
        user_profile['weight'] = (user_profile['rating'] - user_profile['rating'].min()) / \
                                (user_profile['rating'].max() - user_profile['rating'].min())

        # Calculate weighted average of each feature
        weighted_features = {}
        for feature in self.feature_columns:
            weighted_features[feature] = np.average(user_profile[feature], weights=user_profile['weight'])

        # Create a DataFrame with the user profile
        user_profile_df = pd.DataFrame([weighted_features])

        # Find tracks in Spotify data that are not in the user's history
        history_track_ids = set(self.merged_data['track_id'].unique())
        new_tracks = self.spotify_data[~self.spotify_data['track_id'].isin(history_track_ids)].copy()

        if len(new_tracks) == 0:
            print("No new tracks found in Spotify data.")
            return None

        # Calculate similarity between user profile and new tracks
        user_features = user_profile_df[self.feature_columns].values
        track_features = new_tracks[self.feature_columns].values

        # Calculate cosine similarity
        similarities = cosine_similarity(user_features, track_features)[0]

        # Add similarity scores to new_tracks
        new_tracks['similarity'] = similarities

        # Get top n tracks with highest similarity
        top_tracks = new_tracks.sort_values('similarity', ascending=False).head(n)

        # Prepare recommendations
        recommendations = []
        for i, (_, track) in enumerate(top_tracks.iterrows()):
            recommendations.append({
                'rank': i + 1,
                'artist': track['artist_name'],
                'track': track['track_name'],
                'similarity': track['similarity'],
                'track_id': track['track_id']
            })

        return recommendations

    def evaluate_recommendations_with_history(self, recommendations):
        """Validate recommendations by comparing to historical favorites"""
        if self.user_history is None or recommendations is None:
            print("Both user history and recommendations need to be available.")
            return False

        # Get top artists from history
        top_artists = self.user_history.groupby('artistName')['rating'].mean().sort_values(ascending=False).head(10)
        top_artists_list = top_artists.index.tolist()

        # Check how many recommended tracks are from favorite artists
        artist_match_count = 0
        for rec in recommendations:
            if rec['artist'] in top_artists_list:
                artist_match_count += 1

        artist_match_rate = artist_match_count / len(recommendations) * 100
        print(f"Artist match rate: {artist_match_rate:.2f}% of recommendations are from your top artists")

        # Get average audio features for highly rated tracks
        top_tracks = self.merged_data[self.merged_data['rating'] > self.merged_data['rating'].mean()]
        avg_features = top_tracks[self.feature_columns].mean()

        # Compare recommended tracks' features with preferred features
        feature_similarity = []
        for rec in recommendations:
            rec_track = self.spotify_data[self.spotify_data['track_id'] == rec['track_id']]
            if not rec_track.empty:
                rec_features = rec_track[self.feature_columns].values[0]
                similarity = cosine_similarity([avg_features], [rec_features])[0][0]
                feature_similarity.append(similarity)

        avg_similarity = np.mean(feature_similarity) if feature_similarity else 0
        print(f"Average audio feature similarity: {avg_similarity:.4f}")

        return {
            'artist_match_rate': artist_match_rate,
            'feature_similarity': avg_similarity
        }

# Example usage
def run_test_with_history():
    # Initialize the recommender
    recommender = SpotifyHistoryBasedRecommender('history.csv', 'spotify_data.csv')

    # Step 1: Load the data
    if not recommender.load_data():
        return

    # Step 2: Preprocess the listening history
    if not recommender.preprocess_history_data():
        return

    # Step 3: Merge with Spotify data
    if not recommender.merge_with_spotify_data():
        return

    # Step 4: Analyze listening patterns
    influential_features = recommender.analyze_listening_patterns()
    print(f"\nMost influential features: {influential_features}")

    # Step 5: Prepare train/test data
    if not recommender.prepare_train_test_data():
        return

    # Step 6: Build the model
    if not recommender.build_model(num_factors=50):
        return

    # Step 7: Train the model
    recommender.train_model(epochs=20)

    # Step 8: Get recommendations
    print("\nGenerating recommendations based on your listening history...")
    try:
        recommendations = recommender.recommend_new_tracks(n=10)
        if recommendations is None:
            recommendations = recommender.recommend_by_audio_features(n=10)
    except Exception as e:
        print(f"Error generating model-based recommendations: {e}")
        recommendations = recommender.recommend_by_audio_features(n=10)

    # Step 9: Display recommendations
    if recommendations:
        print("\nTop 10 Track Recommendations:")
        for rec in recommendations:
            rank = rec.get('rank', 0)
            score = rec.get('predicted_rating', rec.get('similarity', 0))
            print(f"{rank}. {rec['artist']} - {rec['track']} (Score: {score:.4f})")

        # Step 10: Evaluate recommendations
        print("\nEvaluating recommendations...")
        evaluation = recommender.evaluate_recommendations_with_history(recommendations)
    else:
        print("Could not generate recommendations.")

if __name__ == "__main__":
    run_test_with_history()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score, mean_squared_error
import random

class SpotifyFeatureSimilarityRecommender:
    def __init__(self, spotify_data_path='spotify_data.csv'):
        """Initialize the recommender using only Spotify audio features"""
        self.spotify_data_path = spotify_data_path
        self.spotify_data = None
        self.feature_columns = [
            'danceability', 'energy', 'loudness', 'speechiness',
            'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo'
        ]
        self.scaler = StandardScaler()

    def load_data(self):
        """Load Spotify audio features data"""
        try:
            # Load Spotify data with audio features
            self.spotify_data = pd.read_csv(self.spotify_data_path)
            print(f"Successfully loaded {len(self.spotify_data)} tracks with audio features")

            # Basic data exploration
            print("\nFeature columns statistics:")
            print(self.spotify_data[self.feature_columns].describe())

            # Check for missing values
            missing_values = self.spotify_data[self.feature_columns].isnull().sum()
            if missing_values.sum() > 0:
                print("\nMissing values in features:")
                print(missing_values[missing_values > 0])
                # Handle missing values - fill with mean
                self.spotify_data[self.feature_columns] = self.spotify_data[self.feature_columns].fillna(
                    self.spotify_data[self.feature_columns].mean())
                print("Missing values filled with mean values")
            else:
                print("\nNo missing values in feature columns")

            return True
        except Exception as e:
            print(f"Error loading data: {e}")
            return False

    def preprocess_features(self):
        """Scale and preprocess the audio features"""
        if self.spotify_data is None:
            print("No data loaded. Call load_data() first.")
            return False

        # Scale the features
        self.spotify_data_scaled = self.spotify_data.copy()
        self.spotify_data_scaled[self.feature_columns] = self.scaler.fit_transform(
            self.spotify_data[self.feature_columns])

        print("Audio features scaled successfully")

        # Show correlation between features
        plt.figure(figsize=(12, 10))
        correlation_matrix = self.spotify_data[self.feature_columns].corr()
        sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
        plt.title('Correlation Matrix of Audio Features')
        plt.tight_layout()
        plt.show()

        # Create a PCA model to visualize feature space
        self.visualize_feature_space()

        return True

    def visualize_feature_space(self):
        """Visualize the feature space using dimensionality reduction"""
        # Apply PCA for feature space visualization
        pca = PCA(n_components=2)
        pca_result = pca.fit_transform(self.spotify_data_scaled[self.feature_columns])

        # Create a DataFrame for plotting
        pca_df = pd.DataFrame({
            'PCA1': pca_result[:, 0],
            'PCA2': pca_result[:, 1],
            'track': self.spotify_data_scaled['track_name'],
            'artist': self.spotify_data_scaled['artist_name']
        })

        # Plot PCA results
        plt.figure(figsize=(12, 8))
        sns.scatterplot(data=pca_df, x='PCA1', y='PCA2', alpha=0.6)
        plt.title('PCA Visualization of Audio Features')
        plt.xlabel(f'Principal Component 1 ({pca.explained_variance_ratio_[0]:.2%} variance)')
        plt.ylabel(f'Principal Component 2 ({pca.explained_variance_ratio_[1]:.2%} variance)')
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()

        print(f"Total variance explained by first two principal components: {sum(pca.explained_variance_ratio_):.2%}")

        return pca_df

    def find_optimal_clusters(self, max_clusters=15):
        """Find the optimal number of clusters for KMeans using silhouette scores"""
        X = self.spotify_data_scaled[self.feature_columns].values
        silhouette_scores = []

        for n_clusters in range(2, max_clusters + 1):
            clusterer = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
            cluster_labels = clusterer.fit_predict(X)
            silhouette_avg = silhouette_score(X, cluster_labels)
            silhouette_scores.append(silhouette_avg)
            print(f"For n_clusters = {n_clusters}, the silhouette score is {silhouette_avg:.4f}")

        # Plot silhouette scores
        plt.figure(figsize=(10, 6))
        plt.plot(range(2, max_clusters + 1), silhouette_scores, 'o-', linewidth=2)
        plt.xlabel('Number of Clusters')
        plt.ylabel('Silhouette Score')
        plt.title('Silhouette Score for Different Cluster Counts')
        plt.grid(True, alpha=0.3)
        plt.show()

        # Return the optimal number of clusters
        optimal_clusters = silhouette_scores.index(max(silhouette_scores)) + 2
        print(f"Optimal number of clusters: {optimal_clusters}")
        return optimal_clusters

    def cluster_songs(self, n_clusters=None):
        """Cluster songs based on audio features"""
        if n_clusters is None:
            n_clusters = self.find_optimal_clusters()

        # Apply KMeans clustering
        kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
        self.spotify_data_scaled['cluster'] = kmeans.fit_predict(
            self.spotify_data_scaled[self.feature_columns])

        # Analyze the clusters
        cluster_centers = pd.DataFrame(
            kmeans.cluster_centers_,
            columns=self.feature_columns
        )

        # Plot cluster centers
        plt.figure(figsize=(15, 10))
        ax = sns.heatmap(
            cluster_centers,
            annot=True,
            cmap='viridis',
            fmt=".2f"
        )
        plt.title('Cluster Centers (Standardized Audio Features)')
        plt.xlabel('Audio Features')
        plt.ylabel('Cluster')
        plt.tight_layout()
        plt.show()

        # Count songs per cluster
        cluster_counts = self.spotify_data_scaled['cluster'].value_counts().sort_index()

        # Plot cluster sizes
        plt.figure(figsize=(10, 6))
        cluster_counts.plot(kind='bar')
        plt.title('Number of Songs per Cluster')
        plt.xlabel('Cluster')
        plt.ylabel('Number of Songs')
        plt.tight_layout()
        plt.show()

        print("\nCluster analysis completed")
        return self.spotify_data_scaled

    def calculate_similarity_matrix(self, metric='cosine'):
        """Calculate similarity matrix between all songs"""
        if self.spotify_data_scaled is None:
            print("Data not preprocessed. Call preprocess_features() first.")
            return None

        X = self.spotify_data_scaled[self.feature_columns].values

        if metric == 'cosine':
            # For cosine similarity, higher values (closer to 1) indicate more similarity
            similarity_matrix = cosine_similarity(X)
        elif metric == 'euclidean':
            # For euclidean distance, lower values indicate more similarity
            # We convert to similarity by taking 1 / (1 + distance)
            distances = euclidean_distances(X)
            similarity_matrix = 1 / (1 + distances)
        else:
            print(f"Unsupported metric: {metric}. Using cosine similarity instead.")
            similarity_matrix = cosine_similarity(X)

        print(f"Calculated {metric} similarity matrix of shape {similarity_matrix.shape}")
        return similarity_matrix

    def find_similar_songs(self, song_index, similarity_matrix=None, n=10):
        """Find songs similar to the reference song at song_index"""
        if self.spotify_data_scaled is None:
            print("Data not preprocessed. Call preprocess_features() first.")
            return None

        if similarity_matrix is None:
            similarity_matrix = self.calculate_similarity_matrix()

        # Get the similarities for the reference song
        song_similarities = similarity_matrix[song_index]

        # Get top n similar songs (excluding the reference song itself)
        similar_indices = np.argsort(song_similarities)[::-1][1:n+1]

        # Prepare results
        similar_songs = []
        for i, idx in enumerate(similar_indices):
            similar_songs.append({
                'rank': i + 1,
                'artist': self.spotify_data_scaled.iloc[idx]['artist_name'],
                'track': self.spotify_data_scaled.iloc[idx]['track_name'],
                'similarity': song_similarities[idx],
                'track_id': self.spotify_data_scaled.iloc[idx]['track_id']
            })

        print(f"\nSimilar songs to '{self.spotify_data_scaled.iloc[song_index]['track_name']}' by '{self.spotify_data_scaled.iloc[song_index]['artist_name']}':")

        return similar_songs

    def evaluate_similarity_model(self, test_size=50, n=10):
        """Evaluate the similarity model with random test cases"""
        if self.spotify_data_scaled is None:
            print("Data not preprocessed. Call preprocess_features() first.")
            return None

        # Calculate similarity matrix once
        print("Calculating similarity matrix for evaluation...")
        similarity_matrix = self.calculate_similarity_matrix()

        # Select random songs as test cases
        data_size = len(self.spotify_data_scaled)
        test_indices = random.sample(range(data_size), min(test_size, data_size))

        # Prepare metrics storage
        rmse_values = []
        mae_values = []
        cluster_consistency = []

        for idx in test_indices:
            reference_song = self.spotify_data_scaled.iloc[idx]
            reference_features = reference_song[self.feature_columns].values
            reference_cluster = reference_song['cluster']

            # Get similar songs
            similar_songs = self.find_similar_songs(idx, similarity_matrix, n)

            # Extract features and clusters of similar songs
            similar_indices = [self.spotify_data_scaled.index[self.spotify_data_scaled['track_id'] == song['track_id']].tolist()[0]
                               for song in similar_songs]
            similar_features = self.spotify_data_scaled.iloc[similar_indices][self.feature_columns].values
            similar_clusters = self.spotify_data_scaled.iloc[similar_indices]['cluster'].values

            # Calculate feature differences
            feature_diffs = []
            for i in range(len(similar_songs)):
                # MSE between reference song and similar song
                mse = mean_squared_error(reference_features, similar_features[i])
                feature_diffs.append(mse)

            # Calculate metrics
            rmse = np.sqrt(np.mean(feature_diffs))
            mae = np.mean(np.abs(similar_features - reference_features))

            # Calculate cluster consistency (percentage of similar songs in same cluster)
            same_cluster = np.sum(similar_clusters == reference_cluster)
            cluster_consistency_rate = same_cluster / len(similar_clusters)

            # Store metrics
            rmse_values.append(rmse)
            mae_values.append(mae)
            cluster_consistency.append(cluster_consistency_rate)

        # Calculate average metrics
        avg_rmse = np.mean(rmse_values)
        avg_mae = np.mean(mae_values)
        avg_cluster_consistency = np.mean(cluster_consistency)

        print("\nSimilarity Model Evaluation:")
        print(f"RMSE (test data) = {avg_rmse}")
        print(f"Test MSE: {avg_rmse**2}")
        print(f"MAE = {avg_mae}")
        print(f"Cluster Consistency = {avg_cluster_consistency:.2%}")

        # Plot metric distributions
        fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))

        ax1.hist(rmse_values, bins=10, alpha=0.7)
        ax1.set_title('RMSE Distribution')
        ax1.set_xlabel('RMSE')
        ax1.set_ylabel('Frequency')
        ax1.axvline(avg_rmse, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {avg_rmse:.4f}')
        ax1.legend()

        ax2.hist(mae_values, bins=10, alpha=0.7)
        ax2.set_title('MAE Distribution')
        ax2.set_xlabel('MAE')
        ax2.set_ylabel('Frequency')
        ax2.axvline(avg_mae, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {avg_mae:.4f}')
        ax2.legend()

        ax3.hist(cluster_consistency, bins=10, alpha=0.7)
        ax3.set_title('Cluster Consistency Distribution')
        ax3.set_xlabel('Consistency Rate')
        ax3.set_ylabel('Frequency')
        ax3.axvline(avg_cluster_consistency, color='red', linestyle='dashed', linewidth=2,
                   label=f'Mean: {avg_cluster_consistency:.2%}')
        ax3.legend()

        plt.tight_layout()
        plt.show()

        return {
            'rmse': avg_rmse,
            'mse': avg_rmse**2,
            'mae': avg_mae,
            'cluster_consistency': avg_cluster_consistency,
            'raw_rmse_values': rmse_values,
            'raw_mae_values': mae_values,
            'raw_cluster_consistency': cluster_consistency
        }

    def recommend_by_track_name(self, artist_name, track_name, n=10):
        """Find similar songs to a specified track by name"""
        if self.spotify_data_scaled is None:
            print("Data not preprocessed. Call preprocess_features() first.")
            return None

        # Find the track in the dataset
        track_matches = self.spotify_data_scaled[
            (self.spotify_data_scaled['artist_name'].str.lower() == artist_name.lower()) &
            (self.spotify_data_scaled['track_name'].str.lower() == track_name.lower())
        ]

        if len(track_matches) == 0:
            print(f"Track '{track_name}' by '{artist_name}' not found in the dataset.")
            # Try a fuzzy match
            print("Trying partial match...")
            track_matches = self.spotify_data_scaled[
                (self.spotify_data_scaled['artist_name'].str.lower().str.contains(artist_name.lower())) &
                (self.spotify_data_scaled['track_name'].str.lower().str.contains(track_name.lower()))
            ]

            if len(track_matches) == 0:
                print("No similar tracks found.")
                return None
            print(f"Found {len(track_matches)} partial matches.")

        # Take the first match if multiple matches exist
        reference_idx = track_matches.index[0]

        # Calculate similarity matrix
        similarity_matrix = self.calculate_similarity_matrix()

        # Find similar songs
        similar_songs = self.find_similar_songs(reference_idx, similarity_matrix, n)

        return similar_songs

# Main execution function
def run_feature_based_similarity():
    # Initialize the recommender
    recommender = SpotifyFeatureSimilarityRecommender('spotify_data.csv')

    # Step 1: Load the data
    if not recommender.load_data():
        return

    # Step 2: Preprocess features
    if not recommender.preprocess_features():
        return

    # Step 3: Cluster songs
    recommender.cluster_songs()

    # Step 4: Evaluate the similarity model
    evaluation_results = recommender.evaluate_similarity_model(test_size=50)

    # Step 5: Try a sample recommendation
    print("\nTrying a sample recommendation...")
    # Here you would replace with a known artist and track from your dataset
    sample_artist = "The Beatles"  # Replace with an artist in your dataset
    sample_track = "Hey Jude"      # Replace with a track in your dataset

    try:
        similar_songs = recommender.recommend_by_track_name(sample_artist, sample_track)
        if similar_songs:
            print("\nRecommended songs similar to this track:")
            for song in similar_songs:
                print(f"{song['rank']}. {song['artist']} - {song['track']} (Similarity: {song['similarity']:.4f})")
    except Exception as e:
        print(f"Error in sample recommendation: {e}")
        print("Try with a different track that exists in your dataset")

    return evaluation_results

if __name__ == "__main__":
    evaluation_results = run_feature_based_similarity()
    if evaluation_results:
        print("\nFinal Evaluation Results:")
        print(f"RMSE: {evaluation_results['rmse']:.4f}")
        print(f"MSE: {evaluation_results['mse']:.4f}")
        print(f"MAE: {evaluation_results['mae']:.4f}")
        print(f"Cluster Consistency: {evaluation_results['cluster_consistency']:.2%}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import mean_squared_error
import random

# Initialize style
plt.style.use('ggplot')
sns.set_palette("husl")

class SpotifyFeatureSimilarityRecommender:
    def __init__(self, spotify_data_path='spotify_data.csv'):
        """Initialize the recommender using only Spotify audio features"""
        self.spotify_data_path = spotify_data_path
        self.spotify_data = None
        self.feature_columns = [
            'danceability', 'energy', 'loudness', 'speechiness',
            'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo'
        ]
        self.scaler = MinMaxScaler()
        self.similarity_matrix = None

    def load_and_prepare_data(self):
        """Load and prepare Spotify data with visualization"""
        try:
            # Load Spotify data with audio features
            self.spotify_data = pd.read_csv(self.spotify_data_path)
            print(f"Successfully loaded {len(self.spotify_data)} tracks with audio features")

            # Check for missing values
            missing_values = self.spotify_data[self.feature_columns].isnull().sum()
            if missing_values.sum() > 0:
                print("\nMissing values in features:")
                print(missing_values[missing_values > 0])
                # Handle missing values - fill with mean
                self.spotify_data[self.feature_columns] = self.spotify_data[self.feature_columns].fillna(
                    self.spotify_data[self.feature_columns].mean())
                print("Missing values filled with mean values")
            else:
                print("\nNo missing values in feature columns")

            # Feature normalization using MinMaxScaler
            self.spotify_data[self.feature_columns] = self.scaler.fit_transform(
                self.spotify_data[self.feature_columns])

            # Feature correlation visualization
            plt.figure(figsize=(10, 8))
            sns.heatmap(self.spotify_data[self.feature_columns].corr(), annot=True, cmap='coolwarm')
            plt.title("Audio Feature Correlations")
            plt.show()

            return True
        except Exception as e:
            print(f"Error loading data: {e}")
            return False

    def calculate_song_similarity(self, n_neighbors=10):
        """Calculate similar songs using KNN and audio features"""
        if self.spotify_data is None:
            print("No data loaded. Call load_and_prepare_data() first.")
            return False

        # KNN model for similarity calculation
        knn = NearestNeighbors(n_neighbors=n_neighbors+1, metric='cosine')  # +1 to include the song itself
        knn.fit(self.spotify_data[self.feature_columns])

        # Get distances and indices for all songs
        distances, indices = knn.kneighbors(self.spotify_data[self.feature_columns])

        # Convert distance to similarity (1 - distance)
        self.similarity_matrix = 1 - distances

        # Visualize similarity distribution
        plt.figure(figsize=(10, 6))
        sns.histplot(self.similarity_matrix.flatten(), bins=30, kde=True)
        plt.title("Song Similarity Distribution")
        plt.xlabel("Cosine Similarity Score")
        plt.ylabel("Frequency")
        plt.show()

        print(f"Calculated similarity matrix with shape {self.similarity_matrix.shape}")
        return indices, distances

    def get_similar_songs(self, track_id=None, track_index=None, n=10):
        """Get most similar songs for a given track by ID or index"""
        if self.similarity_matrix is None:
            print("Similarity matrix not calculated. Call calculate_song_similarity() first.")
            return None

        # Determine track index
        if track_index is None and track_id is not None:
            try:
                track_index = self.spotify_data.index[self.spotify_data['track_id'] == track_id][0]
            except IndexError:
                print(f"Track ID {track_id} not found in dataset")
                return None
        elif track_index is None:
            # Use a random track if neither index nor ID is provided
            track_index = np.random.randint(0, len(self.spotify_data))

        # Get reference track info
        reference_track = self.spotify_data.iloc[track_index]
        print(f"\nFinding songs similar to: {reference_track['track_name']} by {reference_track['artist_name']}")

        # Get similar songs from similarity matrix
        # Sort in descending order of similarity and exclude the first one (which is the track itself)
        similar_indices = np.argsort(-self.similarity_matrix[track_index])[1:n+1]
        similar_scores = self.similarity_matrix[track_index][similar_indices]

        # Prepare results
        similar_songs = []
        for idx, score in zip(similar_indices, similar_scores):
            similar_songs.append({
                'track': self.spotify_data.iloc[idx]['track_name'],
                'artist': self.spotify_data.iloc[idx]['artist_name'],
                'similarity': score,
                'track_id': self.spotify_data.iloc[idx]['track_id']
            })

        # Visualize top similar songs
        plt.figure(figsize=(12, 6))
        similar_df = pd.DataFrame(similar_songs)
        sns.barplot(
            x='similarity',
            y='track',
            data=similar_df,
            palette='viridis'
        )
        plt.title(f"Songs Similar to: {reference_track['track_name']}")
        plt.xlabel("Similarity Score")
        plt.ylabel("")
        plt.tight_layout()
        plt.show()

        return similar_songs

    def evaluate_similarity_model(self, test_size=50):
        """Evaluate the similarity model with random test tracks"""
        if self.similarity_matrix is None:
            print("Similarity matrix not calculated. Call calculate_song_similarity() first.")
            return None

        # Select random tracks for evaluation
        test_indices = np.random.choice(len(self.spotify_data), min(test_size, len(self.spotify_data)), replace=False)

        # Collect similarity scores and calculate metrics
        all_similarities = []
        rmse_values = []
        mse_values = []

        for idx in test_indices:
            # Get reference track features
            reference_features = self.spotify_data.iloc[idx][self.feature_columns].values

            # Get similar tracks
            similar_indices = np.argsort(-self.similarity_matrix[idx])[1:6]  # Top 5 similar songs
            similar_scores = self.similarity_matrix[idx][similar_indices]

            # Collect similarity scores for distribution analysis
            all_similarities.extend(similar_scores)

            # Calculate feature differences for error metrics
            for similar_idx in similar_indices:
                similar_features = self.spotify_data.iloc[similar_idx][self.feature_columns].values
                mse = mean_squared_error(reference_features, similar_features)
                rmse = np.sqrt(mse)
                rmse_values.append(rmse)
                mse_values.append(mse)

        # Calculate average metrics
        avg_similarity = np.mean(all_similarities)
        avg_rmse = np.mean(rmse_values)
        avg_mse = np.mean(mse_values)

        # Plot evaluation results
        plt.figure(figsize=(15, 5))

        plt.subplot(1, 3, 1)
        sns.histplot(all_similarities, bins=20, kde=True)
        plt.title("Similarity Scores Distribution")
        plt.xlabel("Similarity Score")

        plt.subplot(1, 3, 2)
        sns.histplot(rmse_values, bins=20, kde=True)
        plt.title("RMSE Distribution")
        plt.xlabel("RMSE Value")

        plt.subplot(1, 3, 3)
        sns.boxplot(x=all_similarities)
        plt.title("Similarity Scores Spread")

        plt.tight_layout()
        plt.show()

        print("\nEvaluation Results:")
        print(f"Average similarity score: {avg_similarity:.4f}")
        print(f"RMSE (test data): {avg_rmse:.4f}")
        print(f"Test MSE: {avg_mse:.4f}")

        return {
            'avg_similarity': avg_similarity,
            'rmse': avg_rmse,
            'mse': avg_mse,
            'all_similarities': all_similarities,
            'all_rmse': rmse_values
        }

    def recommend_by_track_name(self, artist_name, track_name, n=10):
        """Find similar songs to a track specified by artist and name"""
        if self.similarity_matrix is None:
            print("Similarity matrix not calculated. Call calculate_song_similarity() first.")
            return None

        # Find the track in the dataset (exact match)
        track_matches = self.spotify_data[
            (self.spotify_data['artist_name'].str.lower() == artist_name.lower()) &
            (self.spotify_data['track_name'].str.lower() == track_name.lower())
        ]

        if len(track_matches) == 0:
            print(f"Track '{track_name}' by '{artist_name}' not found.")
            print("Trying partial match...")

            # Try partial matching
            track_matches = self.spotify_data[
                (self.spotify_data['artist_name'].str.lower().str.contains(artist_name.lower())) &
                (self.spotify_data['track_name'].str.lower().str.contains(track_name.lower()))
            ]

            if len(track_matches) == 0:
                print("No matches found.")
                return None

            print(f"Found {len(track_matches)} partial matches.")

        # Take the first match
        reference_idx = track_matches.index[0]
        return self.get_similar_songs(track_index=reference_idx, n=n)

    def analyze_feature_importance(self):
        """Analyze which features contribute most to similarity"""
        if self.similarity_matrix is None:
            print("Similarity matrix not calculated. Call calculate_song_similarity() first.")
            return None

        # Calculate feature importance for a sample of pairs
        feature_importance = {feature: [] for feature in self.feature_columns}
        num_samples = min(500, len(self.spotify_data))
        sample_indices = np.random.choice(len(self.spotify_data), num_samples, replace=False)

        for idx in sample_indices:
            # Get top 3 similar songs
            similar_indices = np.argsort(-self.similarity_matrix[idx])[1:4]

            for similar_idx in similar_indices:
                # Calculate feature-wise absolute differences
                for feature in self.feature_columns:
                    feature_diff = abs(self.spotify_data.iloc[idx][feature] -
                                       self.spotify_data.iloc[similar_idx][feature])
                    feature_importance[feature].append(feature_diff)

        # Calculate average differences (smaller = more important for similarity)
        avg_diffs = {feature: np.mean(diffs) for feature, diffs in feature_importance.items()}

        # Normalize to get importance (1 - normalized difference)
        max_diff = max(avg_diffs.values())
        min_diff = min(avg_diffs.values())
        range_diff = max_diff - min_diff

        if range_diff > 0:
            importance_scores = {feature: 1 - ((diff - min_diff) / range_diff)
                                 for feature, diff in avg_diffs.items()}
        else:
            importance_scores = {feature: 1 for feature in self.feature_columns}

        # Sort features by importance
        sorted_features = sorted(importance_scores.items(), key=lambda x: x[1], reverse=True)

        # Plot feature importance
        plt.figure(figsize=(12, 6))
        features = [f[0] for f in sorted_features]
        scores = [f[1] for f in sorted_features]

        sns.barplot(x=features, y=scores)
        plt.title("Feature Importance for Song Similarity")
        plt.xlabel("Audio Feature")
        plt.ylabel("Importance Score")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

        print("\nFeature Importance for Similarity (higher = more important):")
        for feature, score in sorted_features:
            print(f"{feature}: {score:.4f}")

        return sorted_features

    def compare_recommendation_approaches(self, track_id=None, n=10):
        """Compare different approaches to finding similar songs"""
        if self.similarity_matrix is None:
            print("Similarity matrix not calculated. Call calculate_song_similarity() first.")
            return None

        # Select a track if not provided
        if track_id is None:
            track_id = self.spotify_data.sample(1)['track_id'].iloc[0]

        try:
            track_idx = self.spotify_data.index[self.spotify_data['track_id'] == track_id][0]
        except IndexError:
            print(f"Track ID {track_id} not found in dataset")
            return None

        reference_track = self.spotify_data.iloc[track_idx]
        print(f"\nComparing recommendation approaches for: {reference_track['track_name']} by {reference_track['artist_name']}")

        # Approach 1: Standard KNN method (already calculated in similarity_matrix)
        knn_similar_indices = np.argsort(-self.similarity_matrix[track_idx])[1:n+1]
        knn_similar_scores = self.similarity_matrix[track_idx][knn_similar_indices]

        # Approach 2: Direct cosine similarity calculation
        reference_features = self.spotify_data.iloc[track_idx][self.feature_columns].values.reshape(1, -1)
        cosine_similarities = cosine_similarity(reference_features,
                                               self.spotify_data[self.feature_columns])[0]
        cosine_similar_indices = np.argsort(-cosine_similarities)[1:n+1]
        cosine_similar_scores = cosine_similarities[cosine_similar_indices]

        # Compare how many recommendations overlap
        common_recommendations = set(knn_similar_indices).intersection(set(cosine_similar_indices))
        overlap_percentage = len(common_recommendations) / n * 100

        print(f"Overlap between recommendation approaches: {overlap_percentage:.1f}%")

        # Calculate RMSE between approaches
        approach_rmse = np.sqrt(mean_squared_error(
            self.similarity_matrix[track_idx][knn_similar_indices],
            cosine_similarities[cosine_similar_indices]
        ))

        print(f"RMSE between recommendation approaches: {approach_rmse:.4f}")

        # Return results from standard KNN approach
        return self.get_similar_songs(track_index=track_idx, n=n)

    def evaluate_similarity_model(self, test_size=50, n=10):
        """Evaluate the similarity model with random test cases"""
        if self.spotify_data_scaled is None:
            print("Data not preprocessed. Call preprocess_features() first.")
            return None

        # Calculate similarity matrix once
        print("Calculating similarity matrix for evaluation...")
        similarity_matrix = self.calculate_similarity_matrix()

        # Select random songs as test cases
        data_size = len(self.spotify_data_scaled)
        test_indices = random.sample(range(data_size), min(test_size, data_size))

        # Prepare metrics storage
        rmse_values = []
        mae_values = []
        cluster_consistency = []

        for idx in test_indices:
            reference_song = self.spotify_data_scaled.iloc[idx]
            reference_features = reference_song[self.feature_columns].values
            reference_cluster = reference_song['cluster']

            # Get similar songs
            similar_songs = self.find_similar_songs(idx, similarity_matrix, n)

            # Extract features and clusters of similar songs
            similar_indices = [self.spotify_data_scaled.index[self.spotify_data_scaled['track_id'] == song['track_id']].tolist()[0]
                               for song in similar_songs]
            similar_features = self.spotify_data_scaled.iloc[similar_indices][self.feature_columns].values
            similar_clusters = self.spotify_data_scaled.iloc[similar_indices]['cluster'].values

            # Calculate feature differences
            feature_diffs = []
            for i in range(len(similar_songs)):
                # MSE between reference song and similar song
                mse = mean_squared_error(reference_features, similar_features[i])
                feature_diffs.append(mse)

            # Calculate metrics
            rmse = np.sqrt(np.mean(feature_diffs))
            mae = np.mean(np.abs(similar_features - reference_features))

            # Calculate cluster consistency (percentage of similar songs in same cluster)
            same_cluster = np.sum(similar_clusters == reference_cluster)
            cluster_consistency_rate = same_cluster / len(similar_clusters)

            # Store metrics
            rmse_values.append(rmse)
            mae_values.append(mae)
            cluster_consistency.append(cluster_consistency_rate)

        # Calculate average metrics
        avg_rmse = np.mean(rmse_values)
        avg_mae = np.mean(mae_values)
        avg_cluster_consistency = np.mean(cluster_consistency)

        print("\nSimilarity Model Evaluation:")
        print(f"RMSE (test data) = {avg_rmse}")
        print(f"Test MSE: {avg_rmse**2}")
        print(f"MAE = {avg_mae}")
        print(f"Cluster Consistency = {avg_cluster_consistency:.2%}")

        # Plot metric distributions
        fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))

        ax1.hist(rmse_values, bins=10, alpha=0.7)
        ax1.set_title('RMSE Distribution')
        ax1.set_xlabel('RMSE')
        ax1.set_ylabel('Frequency')
        ax1.axvline(avg_rmse, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {avg_rmse:.4f}')
        ax1.legend()

        ax2.hist(mae_values, bins=10, alpha=0.7)
        ax2.set_title('MAE Distribution')
        ax2.set_xlabel('MAE')
        ax2.set_ylabel('Frequency')
        ax2.axvline(avg_mae, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {avg_mae:.4f}')
        ax2.legend()

        ax3.hist(cluster_consistency, bins=10, alpha=0.7)
        ax3.set_title('Cluster Consistency Distribution')
        ax3.set_xlabel('Consistency Rate')
        ax3.set_ylabel('Frequency')
        ax3.axvline(avg_cluster_consistency, color='red', linestyle='dashed', linewidth=2,
                   label=f'Mean: {avg_cluster_consistency:.2%}')
        ax3.legend()

        plt.tight_layout()
        plt.show()

        return {
            'rmse': avg_rmse,
            'mse': avg_rmse**2,
            'mae': avg_mae,
            'cluster_consistency': avg_cluster_consistency,
            'raw_rmse_values': rmse_values,
            'raw_mae_values': mae_values,
            'raw_cluster_consistency': cluster_consistency
        }

    def recommend_by_track_name(self, artist_name, track_name, n=10):
        """Find similar songs to a specified track by name"""
        if self.spotify_data_scaled is None:
            print("Data not preprocessed. Call preprocess_features() first.")
            return None

        # Find the track in the dataset
        track_matches = self.spotify_data_scaled[
            (self.spotify_data_scaled['artist_name'].str.lower() == artist_name.lower()) &
            (self.spotify_data_scaled['track_name'].str.lower() == track_name.lower())
        ]

        if len(track_matches) == 0:
            print(f"Track '{track_name}' by '{artist_name}' not found in the dataset.")
            # Try a fuzzy match
            print("Trying partial match...")
            track_matches = self.spotify_data_scaled[
                (self.spotify_data_scaled['artist_name'].str.lower().str.contains(artist_name.lower())) &
                (self.spotify_data_scaled['track_name'].str.lower().str.contains(track_name.lower()))
            ]

            if len(track_matches) == 0:
                print("No similar tracks found.")
                return None
            print(f"Found {len(track_matches)} partial matches.")

        # Take the first match if multiple matches exist
        reference_idx = track_matches.index[0]

        # Calculate similarity matrix
        similarity_matrix = self.calculate_similarity_matrix()

        # Find similar songs
        similar_songs = self.find_similar_songs(reference_idx, similarity_matrix, n)

        return similar_songs

 Step 5: Try a sample recommendation
    print("\nTrying a sample recommendation...")
    # Here you would replace with a known artist and track from your dataset
    sample_artist = "The Beatles"  # Replace with an artist in your dataset
    sample_track = "Hey Jude"      # Replace with a track in your dataset

    try:
        similar_songs = recommender.recommend_by_track_name(sample_artist, sample_track)
        if similar_songs:
            print("\nRecommended songs similar to this track:")
            for song in similar_songs:
                print(f"{song['rank']}. {song['artist']} - {song['track']} (Similarity: {song['similarity']:.4f})")
    except Exception as e:
        print(f"Error in sample recommendation: {e}")
        print("Try with a different track that exists in your dataset")

    return evaluation_results

if __name__ == "__main__":
    evaluation_results = run_feature_based_similarity()
    if evaluation_results:
        print("\nFinal Evaluation Results:")
        print(f"RMSE: {evaluation_results['rmse']:.4f}")
        print(f"MSE: {evaluation_results['mse']:.4f}")
        print(f"MAE: {evaluation_results['mae']:.4f}")
        print(f"Cluster Consistency: {evaluation_results['cluster_consistency']:.2%}")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.spatial.distance import cosine

class SpotifyRecommender:
    """
    A feature-based song recommendation system using KNN similarity approach
    """

    def __init__(self):
        self.data = None
        self.audio_features = None
        self.feature_names = None
        self.scaler = MinMaxScaler()
        self.model = None

    def load_data(self, file_path):
        """
        Load Spotify data from CSV file
        """
        self.data = pd.read_csv(file_path)
        print(f"Loaded {len(self.data)} tracks")

        # Display basic information
        print(f"Dataset contains {self.data['artist_name'].nunique()} artists")
        print(f"Dataset shape: {self.data.shape}")

        return self.data

    def preprocess_data(self):
        """
        Preprocess the data by:
        1. Handling missing values
        2. Extracting audio features
        3. Normalizing audio features
        """
        # Define audio features to use for similarity calculation
        self.feature_names = ['danceability', 'energy', 'key', 'loudness', 'mode',
                             'speechiness', 'acousticness', 'instrumentalness',
                             'liveness', 'valence', 'tempo', 'duration_ms']

        # Handle missing values by filling with mean
        for feature in self.feature_names:
            if self.data[feature].isnull().sum() > 0:
                self.data[feature].fillna(self.data[feature].mean(), inplace=True)

        # Extract audio features
        self.audio_features = self.data[self.feature_names]

        # Normalize features
        self.audio_features_scaled = self.scaler.fit_transform(self.audio_features)

        print("Data preprocessing complete")
        return self.audio_features_scaled

    def build_model(self, n_neighbors=5):
        """
        Build KNN model for finding similar songs
        """
        self.model = NearestNeighbors(n_neighbors=n_neighbors,
                                       metric='cosine',
                                       algorithm='brute')
        self.model.fit(self.audio_features_scaled)
        print(f"KNN model built with {n_neighbors} neighbors")
        return self.model

    def get_song_index(self, artist_name, track_name):
        """
        Get index of a song based on artist name and track name
        """
        try:
            return self.data[(self.data['artist_name'] == artist_name) &
                            (self.data['track_name'] == track_name)].index[0]
        except IndexError:
            print(f"'{track_name}' by {artist_name} not found in the dataset")
            return None

    def get_song_by_id(self, track_id):
        """
        Get index of a song based on track_id
        """
        try:
            return self.data[self.data['track_id'] == track_id].index[0]
        except IndexError:
            print(f"Track ID {track_id} not found in the dataset")
            return None

    def recommend_songs(self, song_index, n_recommendations=10):
        """
        Recommend similar songs based on audio features
        """
        if song_index is None:
            return None

        # Get song features
        song_features = self.audio_features_scaled[song_index].reshape(1, -1)

        # Find nearest neighbors
        distances, indices = self.model.kneighbors(song_features, n_neighbors=n_recommendations+1)

        # Convert distances to similarity scores (1 - cosine distance)
        similarities = 1 - distances.flatten()

        # Remove the input song itself (first result)
        indices = indices.flatten()[1:]
        similarities = similarities[1:]

        # Create recommendations dataframe
        recommendations = pd.DataFrame({
            'artist_name': self.data['artist_name'].iloc[indices].values,
            'track_name': self.data['track_name'].iloc[indices].values,
            'track_id': self.data['track_id'].iloc[indices].values,
            'similarity_score': similarities
        })

        return recommendations.sort_values('similarity_score', ascending=False)

    def evaluate_similarity_distribution(self):
        """
        Evaluate the distribution of similarity scores
        """
        # Sample songs to evaluate (for efficiency)
        sample_size = min(500, len(self.data))
        sample_indices = np.random.choice(len(self.data), sample_size, replace=False)

        all_similarities = []

        for idx in sample_indices:
            song_features = self.audio_features_scaled[idx].reshape(1, -1)
            distances, _ = self.model.kneighbors(song_features, n_neighbors=6)
            similarities = 1 - distances.flatten()[1:]  # Skip the song itself
            all_similarities.extend(similarities)

        plt.figure(figsize=(10, 6))
        sns.histplot(all_similarities, kde=True)
        plt.title('Distribution of Similarity Scores')
        plt.xlabel('Similarity Score (1 - Cosine Distance)')
        plt.ylabel('Frequency')
        plt.axvline(np.mean(all_similarities), color='red', linestyle='--',
                   label=f'Mean: {np.mean(all_similarities):.3f}')
        plt.legend()
        plt.show()

        print(f"Average similarity score: {np.mean(all_similarities):.3f}")
        print(f"Min similarity score: {np.min(all_similarities):.3f}")
        print(f"Max similarity score: {np.max(all_similarities):.3f}")

        return all_similarities

    def calculate_feature_importance(self, n_samples=100):
        """
        Calculate feature importance by measuring RMSE when each feature is excluded
        """
        base_errors = []
        feature_errors = {feature: [] for feature in self.feature_names}

        # Sample songs to evaluate
        sample_indices = np.random.choice(len(self.data), n_samples, replace=False)

        for idx in sample_indices:
            # Get similar songs based on all features
            song_features = self.audio_features_scaled[idx].reshape(1, -1)
            _, base_indices = self.model.kneighbors(song_features, n_neighbors=6)
            base_indices = base_indices.flatten()[1:]  # Skip the song itself

            base_songs = self.audio_features_scaled[base_indices]
            base_error = np.sqrt(mean_squared_error(
                np.tile(song_features, (len(base_indices), 1)),
                base_songs
            ))
            base_errors.append(base_error)

            # For each feature, calculate error when excluding it
            for i, feature in enumerate(self.feature_names):
                # Create a copy of the scaled features without the current feature
                temp_features = np.delete(self.audio_features_scaled, i, axis=1)

                # Build temporary model
                temp_model = NearestNeighbors(n_neighbors=6,
                                             metric='cosine',
                                             algorithm='brute')
                temp_model.fit(temp_features)

                # Get similar songs excluding the feature
                temp_song_features = np.delete(self.audio_features_scaled[idx], i)
                temp_song_features = temp_song_features.reshape(1, -1)
                _, temp_indices = temp_model.kneighbors(temp_song_features, n_neighbors=6)
                temp_indices = temp_indices.flatten()[1:]  # Skip the song itself

                # Calculate error
                temp_songs = self.audio_features_scaled[temp_indices]
                feature_error = np.sqrt(mean_squared_error(
                    np.tile(song_features, (len(temp_indices), 1)),
                    temp_songs
                ))
                feature_errors[feature].append(feature_error)

        # Calculate importance as the difference in error when removing the feature
        importances = {}
        for feature in self.feature_names:
            importance = np.mean(feature_errors[feature]) - np.mean(base_errors)
            importances[feature] = importance

        # Normalize importances to sum to 1
        total = sum(abs(value) for value in importances.values())
        if total > 0:
            importances = {k: abs(v)/total for k, v in importances.items()}

        # Sort by importance
        importances = dict(sorted(importances.items(), key=lambda x: x[1], reverse=True))

        # Plot feature importances
        plt.figure(figsize=(12, 6))
        plt.bar(importances.keys(), importances.values())
        plt.title('Feature Importance')
        plt.xlabel('Features')
        plt.ylabel('Normalized Importance')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()

        return importances

    def calculate_rmse_and_mse(self, n_samples=100):
        """
        Calculate RMSE and MSE between a song and its recommendations
        """
        rmse_values = []
        mse_values = []

        # Sample songs to evaluate
        sample_indices = np.random.choice(len(self.data), n_samples, replace=False)

        for idx in sample_indices:
            # Get similar songs
            song_features = self.audio_features_scaled[idx].reshape(1, -1)
            _, indices = self.model.kneighbors(song_features, n_neighbors=6)
            indices = indices.flatten()[1:]  # Skip the song itself

            similar_songs = self.audio_features_scaled[indices]

            # Calculate MSE and RMSE
            mse = mean_squared_error(
                np.tile(song_features, (len(indices), 1)),
                similar_songs
            )
            rmse = np.sqrt(mse)

            mse_values.append(mse)
            rmse_values.append(rmse)

        print(f"Average MSE: {np.mean(mse_values):.4f}")
        print(f"Average RMSE: {np.mean(rmse_values):.4f}")

        return np.mean(rmse_values), np.mean(mse_values)

# Demonstration of using the recommender system
def main():
    # Initialize the recommender
    recommender = SpotifyRecommender()

    # Load data
    data = recommender.load_data('spotify_data.csv')

    # Preprocess data
    features_scaled = recommender.preprocess_data()

    # Build the model
    model = recommender.build_model(n_neighbors=10)

    # Example: Get recommendations for a song
    song_index = recommender.get_song_index('Jason Mraz', 'I Won\'t Give Up')
    if song_index is not None:
        print(f"\nRecommendations for 'I Won't Give Up' by Jason Mraz:")
        recommendations = recommender.recommend_songs(song_index, n_recommendations=5)
        print(recommendations)

    # Evaluate the model
    print("\nEvaluating model performance:")
    rmse, mse = recommender.calculate_rmse_and_mse(n_samples=50)

    # Analyze feature importance
    print("\nAnalyzing feature importance:")
    importances = recommender.calculate_feature_importance(n_samples=50)

    # Evaluate similarity distribution
    print("\nEvaluating similarity distribution:")
    similarities = recommender.evaluate_similarity_distribution()

    return recommender

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.initializers import HeNormal, Zeros
import matplotlib.pyplot as plt
import tensorflow as tf
import gc
import os
from tqdm import tqdm

class StableSpotifyRecommender:
    def __init__(self, spotify_data_path='spotify_data.csv', chunk_size=100000, sample_size=100000):
        """Initialize the recommender with improved numerical stability"""
        self.spotify_data_path = spotify_data_path
        self.chunk_size = chunk_size
        self.sample_size = sample_size
        self.model = None
        self.feature_columns = [
            'danceability', 'energy', 'loudness', 'speechiness',
            'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo'
        ]
        self.track_id_mapping = {}
        self.artist_name_mapping = {}
        self.scaler = StandardScaler()  # StandardScaler better for numerical stability

        # Set numeric precision to float32 for better stability
        tf.keras.backend.set_floatx('float32')

        # Configure TensorFlow for better numerical stability
        tf.config.optimizer.set_jit(False)  # Disable XLA to avoid NaN issues

        # Enable memory growth for GPU if available
        gpus = tf.config.experimental.list_physical_devices('GPU')
        if gpus:
            try:
                for gpu in gpus:
                    tf.config.experimental.set_memory_growth(gpu, True)
                print(f"Memory growth enabled on {len(gpus)} GPU(s)")
            except RuntimeError as e:
                print(f"Error setting memory growth: {e}")

    def count_dataset_size(self):
        """Count the total number of records in the dataset without loading it all"""
        total_rows = 0
        for chunk in pd.read_csv(self.spotify_data_path, chunksize=self.chunk_size):
            total_rows += len(chunk)
        print(f"Total rows in dataset: {total_rows}")
        return total_rows

    def create_track_mapping(self):
        """Create a mapping for track IDs from a sample"""
        print("Creating track ID mapping from sample...")
        track_ids = set()
        artist_names = set()

        # Read chunks until we have enough samples
        rows_processed = 0
        for chunk in tqdm(pd.read_csv(self.spotify_data_path, chunksize=self.chunk_size)):
            # Update counts
            track_ids.update(chunk['track_id'].unique())
            if 'artist_name' in chunk.columns:
                artist_names.update(chunk['artist_name'].dropna().unique())

            rows_processed += len(chunk)
            if rows_processed >= self.sample_size:
                break

        # Create mapping dictionaries starting from 1 (reserve 0 for padding)
        self.track_id_mapping = {track_id: i+1 for i, track_id in enumerate(track_ids)}
        self.num_tracks = len(self.track_id_mapping) + 1  # +1 for padding/unknown

        if artist_names:
            self.artist_name_mapping = {artist: i+1 for i, artist in enumerate(artist_names)}
            self.num_artists = len(self.artist_name_mapping) + 1  # +1 for padding/unknown

        print(f"Created mappings for {self.num_tracks-1} unique tracks")
        if artist_names:
            print(f"Created mappings for {self.num_artists-1} unique artists")

        return True

    def process_sample_data(self, output_path='sample_data.csv'):
        """Process a sample of the data with improved numerical handling"""
        print(f"Processing {self.sample_size} sample rows...")

        # Create track mappings first
        self.create_track_mapping()

        # Read and process data in chunks until we have enough samples
        processed_rows = 0
        sample_data = []

        for chunk in tqdm(pd.read_csv(self.spotify_data_path, chunksize=self.chunk_size)):
            # Process chunk
            processed_chunk = self.process_chunk(chunk)

            # Determine how many rows to take from this chunk
            rows_to_take = min(len(processed_chunk), self.sample_size - processed_rows)
            if rows_to_take <= 0:
                break

            # Add to sample data
            sample_data.append(processed_chunk.iloc[:rows_to_take])
            processed_rows += rows_to_take

            # Check if we have enough data
            if processed_rows >= self.sample_size:
                break

            # Clean up memory
            del processed_chunk
            gc.collect()

        # Combine all samples
        combined_sample = pd.concat(sample_data)

        # Save to CSV for simplicity and stability
        combined_sample.to_csv(output_path, index=False)

        print(f"Sample processing complete. {len(combined_sample)} rows processed and saved to {output_path}")

        # Clean up memory
        del sample_data, combined_sample
        gc.collect()

        return output_path

    def process_chunk(self, chunk):
        """Process a single chunk of data with improved numerical handling"""
        # Create a copy to avoid modifying the original
        processed = chunk.copy()

        # Handle missing values in audio features
        for feature in self.feature_columns:
            if feature in processed.columns:
                # Use median for missing values
                median_value = processed[feature].median()
                processed[feature] = processed[feature].fillna(median_value)

                # Clip extreme values to prevent instability
                q1 = processed[feature].quantile(0.01)
                q99 = processed[feature].quantile(0.99)
                processed[feature] = processed[feature].clip(q1, q99)

        # Scale audio features if present
        if all(feature in processed.columns for feature in self.feature_columns):
            # Fit scaler if not already fit
            if not hasattr(self.scaler, 'mean_'):
                self.scaler.fit(processed[self.feature_columns])

            processed[self.feature_columns] = self.scaler.transform(processed[self.feature_columns])

        # Map track IDs using our custom mapping
        processed['track_id_encoded'] = processed['track_id'].map(
            lambda x: self.track_id_mapping.get(x, 0)  # 0 for unknown tracks
        )

        # Map artist names if present
        if 'artist_name' in processed.columns:
            processed['artist_id_encoded'] = processed['artist_name'].map(
                lambda x: self.artist_name_mapping.get(x, 0) if pd.notna(x) else 0
            )

        # Create/normalize ratings
        if 'rating' in processed.columns:
            # Normalize existing ratings to [0, 1] for better stability
            processed['rating'] = (processed['rating'] - 1) / 4.0
        else:
            # Create ratings based on popularity
            if 'popularity' in processed.columns:
                processed['rating'] = processed['popularity'] / 100.0
            else:
                # Random ratings in normalized range
                processed['rating'] = np.random.uniform(0, 1, size=len(processed))

        # Ensure track_id_encoded is integer type
        processed['track_id_encoded'] = processed['track_id_encoded'].astype(np.int32)

        return processed

    def prepare_train_test_split(self, data_path, test_size=0.2, val_size=0.1):
        """Create training, validation and test sets"""
        print("Preparing train/validation/test split...")

        # Load sample data
        df = pd.read_csv(data_path)

        # Split into train, validation, and test sets
        train_val, test = train_test_split(df, test_size=test_size, random_state=42)
        train, val = train_test_split(train_val, test_size=val_size, random_state=42)

        print(f"Train set: {len(train)} rows")
        print(f"Validation set: {len(val)} rows")
        print(f"Test set: {len(test)} rows")

        # Extract features and targets
        X_train = train['track_id_encoded'].values
        y_train = train['rating'].values

        X_val = val['track_id_encoded'].values
        y_val = val['rating'].values

        X_test = test['track_id_encoded'].values
        y_test = test['rating'].values

        return (X_train, y_train), (X_val, y_val), (X_test, y_test)

    def build_model(self, embedding_dim=32):
        """Build a numerically stable neural network model"""
        # Input layer
        track_input = Input(shape=(1,), dtype=tf.int32, name='track-input')

        # Embedding layer with specific initialization for stability
        track_embedding = Embedding(
            self.num_tracks,
            embedding_dim,
            embeddings_initializer=HeNormal(seed=42),
            embeddings_regularizer=tf.keras.regularizers.l2(1e-6),
            name='track-embedding'
        )(track_input)
        track_vec = Flatten(name='track-flatten')(track_embedding)

        # Dense layers with careful initialization and small sizes
        dense_1 = Dense(
            32,
            activation='relu',
            kernel_initializer=HeNormal(seed=43),
            bias_initializer=Zeros(),
            kernel_regularizer=tf.keras.regularizers.l2(1e-6),
            name='dense-1'
        )(track_vec)
        dropout_1 = Dropout(0.2, name='dropout-1')(dense_1)

        # Output layer - linear activation for regression
        output = Dense(
            1,
            activation='linear',
            kernel_initializer='glorot_normal',
            bias_initializer=Zeros(),
            name='rating'
        )(dropout_1)

        # Create and compile model
        model = Model(inputs=track_input, outputs=output)

        # Use a smaller learning rate for stability
        model.compile(
            optimizer=Adam(learning_rate=0.001, epsilon=1e-7),
            loss='mse',
            metrics=['mae']
        )

        self.model = model
        print("Model built successfully:")
        model.summary()
        return model

    def train_model(self, train_data, val_data, epochs=15, batch_size=512):
        """Train the model with numerical stability measures"""
        if self.model is None:
            print("No model created. Call build_model() first.")
            return None

        X_train, y_train = train_data
        X_val, y_val = val_data

        # Create callbacks
        callbacks = [
            EarlyStopping(
                monitor='val_loss',
                patience=5,
                restore_best_weights=True,
                verbose=1
            ),
            ReduceLROnPlateau(
                monitor='val_loss',
                factor=0.5,
                patience=3,
                min_lr=0.0001,
                verbose=1
            )
        ]

        # Train the model
        history = self.model.fit(
            X_train, y_train,
            batch_size=batch_size,
            epochs=epochs,
            validation_data=(X_val, y_val),
            callbacks=callbacks,
            verbose=1
        )

        # Plot training history
        plt.figure(figsize=(12, 5))

        plt.subplot(1, 2, 1)
        plt.plot(history.history['loss'], label='Train Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.title('Model Loss')
        plt.ylabel('Loss')
        plt.xlabel('Epoch')
        plt.legend()

        plt.subplot(1, 2, 2)
        plt.plot(history.history['mae'], label='Train MAE')
        plt.plot(history.history['val_mae'], label='Validation MAE')
        plt.title('Model MAE')
        plt.ylabel('MAE')
        plt.xlabel('Epoch')
        plt.legend()

        plt.tight_layout()
        plt.savefig('model_training_history.png')

        return history

    def evaluate_model(self, test_data):
        """Evaluate the model on test data"""
        if self.model is None:
            print("No model available. Train the model first.")
            return None

        X_test, y_test = test_data

        # Evaluate the model
        results = self.model.evaluate(X_test, y_test, verbose=1)

        print(f"Test Loss (MSE): {results[0]}")
        print(f"Test RMSE: {np.sqrt(results[0])}")
        if len(results) > 1:
            print(f"Test MAE: {results[1]}")

        # Unnormalize metrics for interpretability
        print(f"Unnormalized Test RMSE: {np.sqrt(results[0]) * 4.0}")  # Scale back to 1-5 range
        print(f"Unnormalized Test MAE: {results[1] * 4.0}")  # Scale back to 1-5 range

        return results

    def save_model(self, model_path='spotify_recommender_model'):
        """Save the trained model and necessary preprocessing components"""
        if self.model is None:
            print("No model to save. Train the model first.")
            return False

        # Create directory if it doesn't exist
        if not os.path.exists(model_path):
            os.makedirs(model_path)

        # Save model
        self.model.save(os.path.join(model_path, 'model.keras'))

        # Save components
        import pickle

        # Save track mapping
        with open(os.path.join(model_path, 'track_mapping.pkl'), 'wb') as f:
            pickle.dump(self.track_id_mapping, f)

        # Save artist mapping if exists
        if hasattr(self, 'artist_name_mapping') and self.artist_name_mapping:
            with open(os.path.join(model_path, 'artist_mapping.pkl'), 'wb') as f:
                pickle.dump(self.artist_name_mapping, f)

        # Save scaler
        with open(os.path.join(model_path, 'feature_scaler.pkl'), 'wb') as f:
            pickle.dump(self.scaler, f)

        # Save feature columns
        with open(os.path.join(model_path, 'feature_columns.txt'), 'w') as f:
            f.write('\n'.join(self.feature_columns))

        print(f"Model and components saved to {model_path}")
        return True

    def generate_recommendations(self, data_path, n=10, denormalize=True):
        """Generate recommendations based on the trained model"""
        if self.model is None:
            print("No model available. Train or load a model first.")
            return None

        # Load sample data
        df = pd.read_csv(data_path)

        # Get a sample of tracks to rate
        sample_tracks = df.sample(min(1000, len(df)))
        track_ids = sample_tracks['track_id_encoded'].values

        # Predict ratings
        predictions = self.model.predict(track_ids, verbose=0).flatten()

        # Denormalize predictions if needed
        if denormalize:
            predictions = 1 + (predictions * 4.0)  # Convert back to 1-5 scale

        # Get top n recommendations
        top_indices = predictions.argsort()[-n:][::-1]
        recommended_tracks = []

        for i, idx in enumerate(top_indices):
            track_info = sample_tracks.iloc[idx]
            recommended_tracks.append({
                'rank': i + 1,
                'track_id': track_info['track_id'],
                'artist': track_info.get('artist_name', 'Unknown'),
                'track': track_info.get('track_name', f"Track {track_info['track_id']}"),
                'predicted_rating': predictions[idx]
            })

        return recommended_tracks


def main():
    # Initialize the recommender with smaller sample for faster debugging
    sample_size = 100000  # Smaller sample for testing
    recommender = StableSpotifyRecommender('spotify_data.csv', chunk_size=50000, sample_size=sample_size)

    # Process a sample of the data
    sample_data_path = recommender.process_sample_data()

    # Prepare train/test/validation split
    train_data, val_data, test_data = recommender.prepare_train_test_split(sample_data_path)

    # Build model with stability improvements
    recommender.build_model(embedding_dim=32)

    # Train model with conservative parameters
    recommender.train_model(
        train_data,
        val_data,
        epochs=15,
        batch_size=512
    )

    # Evaluate model
    recommender.evaluate_model(test_data)

    # Save model
    recommender.save_model()

    # Generate recommendations
    recommendations = recommender.generate_recommendations(sample_data_path, n=10)

    if recommendations:
        print("\nTop 10 Track Recommendations:")
        for rec in recommendations:
            print(f"{rec['rank']}. {rec['artist']} - {rec['track']} (Predicted Rating: {rec['predicted_rating']:.2f})")


if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.regularizers import l2
import matplotlib.pyplot as plt
import tensorflow as tf
import gc
import os
from tqdm import tqdm

class ImprovedValidationRecommender:
    def __init__(self, spotify_data_path='spotify_data.csv', chunk_size=50000, sample_size=200000):
        """Initialize the recommender with improved validation performance measures"""
        self.spotify_data_path = spotify_data_path
        self.chunk_size = chunk_size
        self.sample_size = sample_size
        self.model = None
        self.feature_columns = [
            'danceability', 'energy', 'loudness', 'speechiness',
            'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo'
        ]
        self.track_id_mapping = {}
        self.artist_name_mapping = {}
        self.scaler = StandardScaler()

        # Set numeric precision to float32
        tf.keras.backend.set_floatx('float32')

        # Configure TensorFlow for better numerical stability
        tf.config.optimizer.set_jit(False)  # Disable XLA to avoid NaN issues

        # Enable memory growth for GPU if available
        gpus = tf.config.experimental.list_physical_devices('GPU')
        if gpus:
            try:
                for gpu in gpus:
                    tf.config.experimental.set_memory_growth(gpu, True)
                print(f"Memory growth enabled on {len(gpus)} GPU(s)")
            except RuntimeError as e:
                print(f"Error setting memory growth: {e}")

    def create_track_mapping(self):
        """Create a mapping for track IDs with artist grouping to prevent data leakage"""
        print("Creating track ID mapping with artist grouping...")
        track_ids = set()
        artist_tracks = {}  # Group tracks by artist

        # Read chunks until we have enough samples
        rows_processed = 0
        for chunk in tqdm(pd.read_csv(self.spotify_data_path, chunksize=self.chunk_size)):
            # Group tracks by artist
            if 'artist_name' in chunk.columns and 'track_id' in chunk.columns:
                for _, row in chunk.iterrows():
                    artist = row.get('artist_name')
                    track = row.get('track_id')

                    if pd.notna(artist) and pd.notna(track):
                        if artist not in artist_tracks:
                            artist_tracks[artist] = []
                        artist_tracks[artist].append(track)

            # Also add individual tracks
            track_ids.update(chunk['track_id'].unique())

            rows_processed += len(chunk)
            if rows_processed >= self.sample_size:
                break

        # Create mapping dictionaries starting from 1 (reserve 0 for padding)
        self.track_id_mapping = {track_id: i+1 for i, track_id in enumerate(track_ids)}
        self.num_tracks = len(self.track_id_mapping) + 1  # +1 for padding/unknown

        # Create artist mapping
        self.artist_name_mapping = {artist: i+1 for i, artist in enumerate(artist_tracks.keys())}
        self.num_artists = len(self.artist_name_mapping) + 1

        # Store artist-track relationships for stratified splitting
        self.artist_tracks = artist_tracks

        print(f"Created mappings for {self.num_tracks-1} unique tracks")
        print(f"Created mappings for {self.num_artists-1} unique artists")

        return True

    def process_sample_data(self, output_path='sample_data.csv'):
        """Process a sample of the data with artist-track grouping"""
        print(f"Processing {self.sample_size} sample rows...")

        # Create track mappings first with artist grouping
        self.create_track_mapping()

        # Read and process data in chunks until we have enough samples
        processed_rows = 0
        sample_data = []

        for chunk in tqdm(pd.read_csv(self.spotify_data_path, chunksize=self.chunk_size)):
            # Process chunk
            processed_chunk = self.process_chunk(chunk)

            # Determine how many rows to take from this chunk
            rows_to_take = min(len(processed_chunk), self.sample_size - processed_rows)
            if rows_to_take <= 0:
                break

            # Add to sample data
            sample_data.append(processed_chunk.iloc[:rows_to_take])
            processed_rows += rows_to_take

            # Check if we have enough data
            if processed_rows >= self.sample_size:
                break

            # Clean up memory
            del processed_chunk
            gc.collect()

        # Combine all samples
        combined_sample = pd.concat(sample_data)

        # Add additional engineered features
        combined_sample = self.add_engineered_features(combined_sample)

        # Save to CSV
        combined_sample.to_csv(output_path, index=False)

        print(f"Sample processing complete. {len(combined_sample)} rows processed and saved to {output_path}")

        # Clean up memory
        del sample_data, combined_sample
        gc.collect()

        return output_path

    def process_chunk(self, chunk):
        """Process a single chunk of data with improved handling"""
        # Create a copy to avoid modifying the original
        processed = chunk.copy()

        # Handle missing values in audio features
        for feature in self.feature_columns:
            if feature in processed.columns:
                # Use median for missing values
                median_value = processed[feature].median()
                processed[feature] = processed[feature].fillna(median_value)

                # Clip extreme values to prevent instability
                q1 = processed[feature].quantile(0.01)
                q99 = processed[feature].quantile(0.99)
                processed[feature] = processed[feature].clip(q1, q99)

        # Scale audio features if present
        if all(feature in processed.columns for feature in self.feature_columns):
            # Fit scaler if not already fit
            if not hasattr(self.scaler, 'mean_'):
                self.scaler.fit(processed[self.feature_columns])

            processed[self.feature_columns] = self.scaler.transform(processed[self.feature_columns])

        # Map track IDs using our custom mapping
        processed['track_id_encoded'] = processed['track_id'].map(
            lambda x: self.track_id_mapping.get(x, 0)  # 0 for unknown tracks
        )

        # Map artist names if present
        if 'artist_name' in processed.columns:
            processed['artist_id_encoded'] = processed['artist_name'].map(
                lambda x: self.artist_name_mapping.get(x, 0) if pd.notna(x) else 0
            )

        # Create/normalize ratings
        if 'rating' in processed.columns:
            # Normalize existing ratings to [0, 1] for better stability
            processed['rating'] = (processed['rating'] - 1) / 4.0
        else:
            # Create ratings based on popularity
            if 'popularity' in processed.columns:
                processed['rating'] = processed['popularity'] / 100.0
            else:
                # Random ratings in normalized range
                processed['rating'] = np.random.uniform(0, 1, size=len(processed))

        # Ensure track_id_encoded is integer type
        processed['track_id_encoded'] = processed['track_id_encoded'].astype(np.int32)

        return processed

    def add_engineered_features(self, df):
        """Add engineered features to improve model performance"""
        # Create interaction features between audio characteristics
        df['energy_loudness'] = df['energy'] * df['loudness']
        df['dance_valence'] = df['danceability'] * df['valence']
        df['acoustic_instrumental'] = df['acousticness'] * df['instrumentalness']

        # Create categorical features from continuous ones
        for feature in self.feature_columns:
            if feature in df.columns:
                # Create binary indicators for high/low values
                df[f'{feature}_high'] = (df[feature] > df[feature].mean()).astype(float)

        # Add popularity bins if available
        if 'popularity' in df.columns:
            df['popularity_binned'] = pd.qcut(
                df['popularity'],
                q=5,
                labels=False,
                duplicates='drop'
            ).astype(float) / 4.0  # Normalize to [0,1]

        return df

    def prepare_stratified_split(self, data_path, test_size=0.2, val_size=0.1):
        """Create training, validation and test sets with artist stratification"""
        print("Preparing artist-stratified train/validation/test split...")

        # Load sample data
        df = pd.read_csv(data_path)

        # Create stratified split - group by artist to prevent data leakage
        artist_groups = df['artist_id_encoded'].unique()

        # Split artist groups
        train_val_artists, test_artists = train_test_split(
            artist_groups, test_size=test_size, random_state=42
        )

        train_artists, val_artists = train_test_split(
            train_val_artists, test_size=val_size/(1-test_size), random_state=42
        )

        # Create data splits based on artist groups
        train = df[df['artist_id_encoded'].isin(train_artists)]
        val = df[df['artist_id_encoded'].isin(val_artists)]
        test = df[df['artist_id_encoded'].isin(test_artists)]

        print(f"Train set: {len(train)} rows, {len(train_artists)} artists")
        print(f"Validation set: {len(val)} rows, {len(val_artists)} artists")
        print(f"Test set: {len(test)} rows, {len(test_artists)} artists")

        # Extract features and targets
        X_train = {
            'track_input': train['track_id_encoded'].values,
            'artist_input': train['artist_id_encoded'].values
        }
        y_train = train['rating'].values

        X_val = {
            'track_input': val['track_id_encoded'].values,
            'artist_input': val['artist_id_encoded'].values
        }
        y_val = val['rating'].values

        X_test = {
            'track_input': test['track_id_encoded'].values,
            'artist_input': test['artist_id_encoded'].values
        }
        y_test = test['rating'].values

        return (X_train, y_train), (X_val, y_val), (X_test, y_test)

    def build_improved_model(self, embedding_dim=16):
        """Build a model with improved validation performance"""
        # Input layers
        track_input = Input(shape=(1,), dtype=tf.int32, name='track_input')
        artist_input = Input(shape=(1,), dtype=tf.int32, name='artist_input')

        # Track embedding with strong regularization
        track_embedding = Embedding(
            self.num_tracks,
            embedding_dim,
            embeddings_initializer='he_normal',
            embeddings_regularizer=l2(1e-4),  # Stronger regularization
            name='track_embedding'
        )(track_input)
        track_vec = Flatten(name='track_flatten')(track_embedding)

        # Artist embedding to capture artist-specific patterns
        artist_embedding = Embedding(
            self.num_artists,
            embedding_dim // 2,  # Smaller dimension for artist
            embeddings_initializer='he_normal',
            embeddings_regularizer=l2(1e-4),
            name='artist_embedding'
        )(artist_input)
        artist_vec = Flatten(name='artist_flatten')(artist_embedding)

        # Concatenate embeddings
        concat = tf.keras.layers.concatenate([track_vec, artist_vec], name='concat_embedding')

        # First dense layer with batch normalization
        dense_1 = Dense(
            32,
            activation='relu',
            kernel_regularizer=l2(1e-4),
            name='dense_1'
        )(concat)
        batch_norm_1 = BatchNormalization(name='batch_norm_1')(dense_1)
        dropout_1 = Dropout(0.4, name='dropout_1')(batch_norm_1)  # Increased dropout

        # Second dense layer with batch normalization
        dense_2 = Dense(
            16,
            activation='relu',
            kernel_regularizer=l2(1e-4),
            name='dense_2'
        )(dropout_1)
        batch_norm_2 = BatchNormalization(name='batch_norm_2')(dense_2)
        dropout_2 = Dropout(0.4, name='dropout_2')(batch_norm_2)  # Increased dropout

        # Output layer - linear activation for regression
        output = Dense(
            1,
            activation='linear',
            kernel_regularizer=l2(1e-4),
            name='rating'
        )(dropout_2)

        # Create and compile model
        model = Model(inputs=[track_input, artist_input], outputs=output)

        # Use a smaller learning rate with conservative optimizer settings
        model.compile(
            optimizer=Adam(learning_rate=0.0005, epsilon=1e-7),  # Lower learning rate
            loss='mse',
            metrics=['mae']
        )

        self.model = model
        print("Improved model built successfully:")
        model.summary()
        return model

    def train_model(self, train_data, val_data, epochs=20, batch_size=256):
        """Train the model with measures to improve validation performance"""
        if self.model is None:
            print("No model created. Call build_improved_model() first.")
            return None

        X_train, y_train = train_data
        X_val, y_val = val_data

        # Create callbacks with more patience to avoid early convergence
        callbacks = [
            EarlyStopping(
                monitor='val_loss',
                patience=7,  # Increased patience
                restore_best_weights=True,
                verbose=1
            ),
            ReduceLROnPlateau(
                monitor='val_loss',
                factor=0.3,  # More aggressive reduction
                patience=3,
                min_lr=0.00001,
                verbose=1
            )
        ]

        # Train the model
        history = self.model.fit(
            X_train, y_train,
            batch_size=batch_size,
            epochs=epochs,
            validation_data=(X_val, y_val),
            callbacks=callbacks,
            verbose=1,
            shuffle=True  # Ensure good shuffling
        )

        # Plot training history
        plt.figure(figsize=(12, 5))

        plt.subplot(1, 2, 1)
        plt.plot(history.history['loss'], label='Train Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.title('Model Loss')
        plt.ylabel('Loss')
        plt.xlabel('Epoch')
        plt.legend()

        plt.subplot(1, 2, 2)
        plt.plot(history.history['mae'], label='Train MAE')
        plt.plot(history.history['val_mae'], label='Validation MAE')
        plt.title('Model MAE')
        plt.ylabel('MAE')
        plt.xlabel('Epoch')
        plt.legend()

        plt.tight_layout()
        plt.savefig('improved_model_training_history.png')

        return history

    def evaluate_model(self, test_data):
        """Evaluate the model on test data"""
        if self.model is None:
            print("No model available. Train the model first.")
            return None

        X_test, y_test = test_data

        # Evaluate the model
        results = self.model.evaluate(X_test, y_test, verbose=1)

        print(f"Test Loss (MSE): {results[0]}")
        print(f"Test RMSE: {np.sqrt(results[0])}")
        if len(results) > 1:
            print(f"Test MAE: {results[1]}")

        # Unnormalize metrics for interpretability
        print(f"Unnormalized Test RMSE: {np.sqrt(results[0]) * 4.0}")  # Scale back to 1-5 range
        print(f"Unnormalized Test MAE: {results[1] * 4.0}")  # Scale back to 1-5 range

        # Visualize predictions vs actual
        predictions = self.model.predict(X_test)

        plt.figure(figsize=(10, 6))
        plt.scatter(y_test, predictions, alpha=0.3)
        plt.plot([0, 1], [0, 1], 'r--')
        plt.xlabel('Actual Rating')
        plt.ylabel('Predicted Rating')
        plt.title('Predictions vs Actual')
        plt.savefig('predictions_vs_actual.png')

        return results

    def save_model(self, model_path='improved_spotify_model'):
        """Save the trained model and necessary preprocessing components"""
        if self.model is None:
            print("No model to save. Train the model first.")
            return False

        # Create directory if it doesn't exist
        if not os.path.exists(model_path):
            os.makedirs(model_path)

        # Save model
        self.model.save(os.path.join(model_path, 'model.keras'))

        # Save components
        import pickle

        # Save track mapping
        with open(os.path.join(model_path, 'track_mapping.pkl'), 'wb') as f:
            pickle.dump(self.track_id_mapping, f)

        # Save artist mapping if exists
        if hasattr(self, 'artist_name_mapping') and self.artist_name_mapping:
            with open(os.path.join(model_path, 'artist_mapping.pkl'), 'wb') as f:
                pickle.dump(self.artist_name_mapping, f)

        # Save scaler
        with open(os.path.join(model_path, 'feature_scaler.pkl'), 'wb') as f:
            pickle.dump(self.scaler, f)

        # Save feature columns
        with open(os.path.join(model_path, 'feature_columns.txt'), 'w') as f:
            f.write('\n'.join(self.feature_columns))

        print(f"Model and components saved to {model_path}")
        return True

    def generate_recommendations(self, data_path, n=10, denormalize=True):
        """Generate recommendations based on the trained model"""
        if self.model is None:
            print("No model available. Train or load a model first.")
            return None

        # Load sample data
        df = pd.read_csv(data_path)

        # Get a sample of tracks to rate
        sample_tracks = df.sample(min(1000, len(df)))

        # Prepare input data
        X_predict = {
            'track_input': sample_tracks['track_id_encoded'].values,
            'artist_input': sample_tracks['artist_id_encoded'].values
        }

        # Predict ratings
        predictions = self.model.predict(X_predict, verbose=0).flatten()

        # Denormalize predictions if needed
        if denormalize:
            predictions = 1 + (predictions * 4.0)  # Convert back to 1-5 scale

        # Get top n recommendations
        top_indices = predictions.argsort()[-n:][::-1]
        recommended_tracks = []

        for i, idx in enumerate(top_indices):
            track_info = sample_tracks.iloc[idx]
            recommended_tracks.append({
                'rank': i + 1,
                'track_id': track_info['track_id'],
                'artist': track_info.get('artist_name', 'Unknown'),
                'track': track_info.get('track_name', f"Track {track_info['track_id']}"),
                'predicted_rating': predictions[idx]
            })

        return recommended_tracks


def main():
    # Initialize the recommender with larger sample for better generalization
    sample_size = 200000
    recommender = ImprovedValidationRecommender('spotify_data.csv', chunk_size=50000, sample_size=sample_size)

    # Process a sample of the data
    sample_data_path = recommender.process_sample_data()

    # Prepare train/test/validation split with artist stratification
    train_data, val_data, test_data = recommender.prepare_stratified_split(sample_data_path)

    # Build improved model
    recommender.build_improved_model(embedding_dim=16)

    # Train model
    recommender.train_model(
        train_data,
        val_data,
        epochs=20,
        batch_size=256
    )

    # Evaluate model
    recommender.evaluate_model(test_data)

    # Save model
    recommender.save_model()

    # Generate recommendations
    recommendations = recommender.generate_recommendations(sample_data_path, n=10)

    if recommendations:
        print("\nTop 10 Track Recommendations:")
        for rec in recommendations:
            print(f"{rec['rank']}. {rec['artist']} - {rec['track']} (Predicted Rating: {rec['predicted_rating']:.2f})")


if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Dropout, BatchNormalization, concatenate
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.regularizers import l2
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
import tensorflow as tf
import gc
import os
from tqdm import tqdm

class ContentBasedMusicRecommender:
    def __init__(self, spotify_data_path='spotify_data.csv', chunk_size=50000, sample_size=200000):
        """Initialize a content-based music recommender that doesn't use ratings"""
        self.spotify_data_path = spotify_data_path
        self.chunk_size = chunk_size
        self.sample_size = sample_size
        self.encoder_model = None
        self.feature_columns = [
            'danceability', 'energy', 'loudness', 'speechiness',
            'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo'
        ]
        self.scaler = StandardScaler()

        # Set numeric precision to float32
        tf.keras.backend.set_floatx('float32')

        # Configure TensorFlow for better numerical stability
        tf.config.optimizer.set_jit(False)  # Disable XLA to avoid NaN issues

        # Enable memory growth for GPU if available
        gpus = tf.config.experimental.list_physical_devices('GPU')
        if gpus:
            try:
                for gpu in gpus:
                    tf.config.experimental.set_memory_growth(gpu, True)
                print(f"Memory growth enabled on {len(gpus)} GPU(s)")
            except RuntimeError as e:
                print(f"Error setting memory growth: {e}")

    def process_sample_data(self, output_path='content_sample_data.csv'):
        """Process a sample of the data for content-based similarity"""
        print(f"Processing {self.sample_size} sample rows...")

        # Read and process data in chunks until we have enough samples
        processed_rows = 0
        sample_data = []

        for chunk in tqdm(pd.read_csv(self.spotify_data_path, chunksize=self.chunk_size)):
            # Process chunk
            processed_chunk = self.process_chunk(chunk)

            # Determine how many rows to take from this chunk
            rows_to_take = min(len(processed_chunk), self.sample_size - processed_rows)
            if rows_to_take <= 0:
                break

            # Add to sample data
            sample_data.append(processed_chunk.iloc[:rows_to_take])
            processed_rows += rows_to_take

            # Check if we have enough data
            if processed_rows >= self.sample_size:
                break

            # Clean up memory
            del processed_chunk
            gc.collect()

        # Combine all samples
        combined_sample = pd.concat(sample_data)

        # Add additional engineered features
        combined_sample = self.add_engineered_features(combined_sample)

        # Save to CSV
        combined_sample.to_csv(output_path, index=False)

        print(f"Sample processing complete. {len(combined_sample)} rows processed and saved to {output_path}")

        # Clean up memory
        del sample_data, combined_sample
        gc.collect()

        return output_path

    def process_chunk(self, chunk):
        """Process a single chunk of data with improved handling"""
        # Create a copy to avoid modifying the original
        processed = chunk.copy()

        # Keep only necessary columns: track_id, artist_name, track_name, and audio features
        keep_columns = ['track_id', 'artist_name', 'track_name'] + self.feature_columns
        processed = processed[[col for col in keep_columns if col in processed.columns]]

        # Handle missing values in audio features
        for feature in self.feature_columns:
            if feature in processed.columns:
                # Use median for missing values
                median_value = processed[feature].median()
                processed[feature] = processed[feature].fillna(median_value)

                # Clip extreme values to prevent instability
                q1 = processed[feature].quantile(0.01)
                q99 = processed[feature].quantile(0.99)
                processed[feature] = processed[feature].clip(q1, q99)

        # Scale audio features if present
        if all(feature in processed.columns for feature in self.feature_columns):
            # Fit scaler if not already fit
            if not hasattr(self.scaler, 'mean_'):
                self.scaler.fit(processed[self.feature_columns])

            processed[self.feature_columns] = self.scaler.transform(processed[self.feature_columns])

        return processed

    def add_engineered_features(self, df):
        """Add engineered features to improve model performance"""
        # Create interaction features between audio characteristics
        df['energy_loudness'] = df['energy'] * df['loudness']
        df['dance_valence'] = df['danceability'] * df['valence']
        df['acoustic_instrumental'] = df['acousticness'] * df['instrumentalness']

        # Create feature to represent electronic vs. acoustic spectrum
        df['electronic_score'] = df['energy'] - df['acousticness']

        # Create feature to represent vocal vs. instrumental spectrum
        df['vocal_score'] = df['speechiness'] - df['instrumentalness']

        # Create mood feature (combination of valence and energy)
        df['mood_score'] = 0.5 * df['valence'] + 0.5 * df['energy']

        # Create danceability intensity feature
        df['dance_intensity'] = df['danceability'] * df['energy']

        # Add genre information if available
        if 'genre' in df.columns:
            # One-hot encode the genre (simplified here)
            genre_dummies = pd.get_dummies(df['genre'], prefix='genre')
            df = pd.concat([df, genre_dummies], axis=1)

        return df

    def prepare_data_for_training(self, data_path):
        """Prepare data for training the autoencoder"""
        print("Preparing data for autoencoder training...")

        # Load sample data
        df = pd.read_csv(data_path)

        # Extract feature columns (original and engineered)
        feature_cols = [col for col in df.columns if col not in ['track_id', 'artist_name', 'track_name']]

        # Create a copy of the features as both input and target for autoencoder
        features = df[feature_cols].values

        # Split into train and validation sets
        X_train, X_val = train_test_split(features, test_size=0.2, random_state=42)

        print(f"Training data shape: {X_train.shape}")
        print(f"Validation data shape: {X_val.shape}")
        print(f"Number of features: {len(feature_cols)}")

        # Save feature column names for later reference
        self.all_feature_columns = feature_cols

        return X_train, X_val, df

    def build_autoencoder(self, input_dim, encoding_dim=32):
        """Build an autoencoder model to create a compact representation of audio features"""
        # Input layer
        input_layer = Input(shape=(input_dim,), name='input')

        # Encoder
        encoded = Dense(64, activation='relu', kernel_regularizer=l2(1e-4), name='encode1')(input_layer)
        encoded = BatchNormalization(name='encode1_bn')(encoded)
        encoded = Dropout(0.3, name='encode1_dropout')(encoded)

        encoded = Dense(encoding_dim, activation='relu', kernel_regularizer=l2(1e-4), name='bottleneck')(encoded)

        # Decoder
        decoded = Dense(64, activation='relu', kernel_regularizer=l2(1e-4), name='decode1')(encoded)
        decoded = BatchNormalization(name='decode1_bn')(decoded)
        decoded = Dropout(0.3, name='decode1_dropout')(decoded)

        decoded = Dense(input_dim, activation='linear', name='output')(decoded)

        # Full autoencoder model
        autoencoder = Model(inputs=input_layer, outputs=decoded)

        # Encoder model for extracting embeddings
        encoder = Model(inputs=input_layer, outputs=encoded)

        # Compile the model
        autoencoder.compile(
            optimizer=Adam(learning_rate=0.001),
            loss='mse'
        )

        self.autoencoder = autoencoder
        self.encoder_model = encoder

        print("Autoencoder built successfully:")
        autoencoder.summary()

        return autoencoder, encoder

    def train_autoencoder(self, X_train, X_val, epochs=30, batch_size=256):
        """Train the autoencoder model"""
        if self.autoencoder is None:
            print("No autoencoder model created. Call build_autoencoder() first.")
            return None

        # Create callbacks
        callbacks = [
            EarlyStopping(
                monitor='val_loss',
                patience=5,
                restore_best_weights=True,
                verbose=1
            ),
            ReduceLROnPlateau(
                monitor='val_loss',
                factor=0.5,
                patience=3,
                min_lr=0.0001,
                verbose=1
            )
        ]

        # Train the model
        history = self.autoencoder.fit(
            X_train, X_train,  # Autoencoder learns to reproduce its input
            batch_size=batch_size,
            epochs=epochs,
            validation_data=(X_val, X_val),
            callbacks=callbacks,
            verbose=1
        )

        # Plot training history
        plt.figure(figsize=(10, 6))
        plt.plot(history.history['loss'], label='Train Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.title('Autoencoder Training Loss')
        plt.ylabel('Loss')
        plt.xlabel('Epoch')
        plt.legend()
        plt.savefig('autoencoder_training_history.png')

        return history

    def create_embeddings(self, data_path):
        """Create embeddings for all tracks using the trained encoder"""
        if self.encoder_model is None:
            print("No encoder model available. Train the autoencoder first.")
            return None

        # Load data
        df = pd.read_csv(data_path)

        # Extract features
        features = df[self.all_feature_columns].values

        # Generate embeddings
        embeddings = self.encoder_model.predict(features)

        # Add embeddings to dataframe
        embedding_df = pd.DataFrame(
            embeddings,
            columns=[f'embedding_{i}' for i in range(embeddings.shape[1])]
        )

        # Add track information
        embedding_df['track_id'] = df['track_id'].values
        if 'artist_name' in df.columns:
            embedding_df['artist_name'] = df['artist_name'].values
        if 'track_name' in df.columns:
            embedding_df['track_name'] = df['track_name'].values

        print(f"Created embeddings with shape: {embeddings.shape}")

        return embedding_df

    def find_similar_tracks(self, track_id, embedding_df, n=10):
        """Find similar tracks based on embedding similarity"""
        # Locate the reference track
        if track_id not in embedding_df['track_id'].values:
            print(f"Track ID {track_id} not found in embeddings.")
            return None

        # Get the track's embedding
        track_embedding = embedding_df[embedding_df['track_id'] == track_id].iloc[0]
        embedding_cols = [col for col in embedding_df.columns if col.startswith('embedding_')]
        track_vector = track_embedding[embedding_cols].values.reshape(1, -1)

        # Calculate similarity to all other tracks
        all_embeddings = embedding_df[embedding_cols].values
        similarities = cosine_similarity(track_vector, all_embeddings)[0]

        # Add similarities to a copy of the dataframe
        results_df = embedding_df.copy()
        results_df['similarity'] = similarities

        # Sort by similarity (excluding the reference track)
        similar_tracks = results_df[results_df['track_id'] != track_id].sort_values(
            'similarity', ascending=False
        ).head(n)

        return similar_tracks

    def recommend_for_playlist(self, track_ids, embedding_df, n=10):
        """Recommend tracks similar to a playlist of tracks"""
        if not all(track_id in embedding_df['track_id'].values for track_id in track_ids):
            missing_tracks = [track_id for track_id in track_ids if track_id not in embedding_df['track_id'].values]
            print(f"Some track IDs not found: {missing_tracks}")
            # Continue with available tracks
            track_ids = [track_id for track_id in track_ids if track_id in embedding_df['track_id'].values]
            if not track_ids:
                return None

        # Get embeddings for the playlist tracks
        embedding_cols = [col for col in embedding_df.columns if col.startswith('embedding_')]
        playlist_embeddings = embedding_df[embedding_df['track_id'].isin(track_ids)][embedding_cols].values

        # Calculate average embedding (simple centroid)
        average_embedding = playlist_embeddings.mean(axis=0).reshape(1, -1)

        # Calculate similarity to all tracks
        all_embeddings = embedding_df[embedding_cols].values
        similarities = cosine_similarity(average_embedding, all_embeddings)[0]

        # Add similarities to a copy of the dataframe
        results_df = embedding_df.copy()
        results_df['similarity'] = similarities

        # Sort by similarity (excluding tracks already in the playlist)
        recommendations = results_df[~results_df['track_id'].isin(track_ids)].sort_values(
            'similarity', ascending=False
        ).head(n)

        return recommendations

    def create_genre_embeddings(self, embedding_df):
        """Create embeddings for genres by averaging track embeddings"""
        if 'genre' not in embedding_df.columns:
            print("No genre information available in embeddings.")
            return None

        # Get embedding columns
        embedding_cols = [col for col in embedding_df.columns if col.startswith('embedding_')]

        # Group by genre and calculate average embeddings
        genre_embeddings = embedding_df.groupby('genre')[embedding_cols].mean()

        return genre_embeddings

    def recommend_by_audio_profile(self, audio_profile, embedding_df, n=10):
        """Recommend tracks based on a specific audio profile"""
        # The audio profile should be a dictionary with audio features as keys
        if not set(audio_profile.keys()).issubset(set(self.feature_columns)):
            print("Invalid audio profile. Keys must be valid audio features.")
            print(f"Valid features: {self.feature_columns}")
            return None

        # Create a feature vector for the profile
        feature_vector = np.zeros((1, len(self.all_feature_columns)))

        # Fill in the provided features
        for i, feature in enumerate(self.all_feature_columns):
            if feature in audio_profile:
                # Scale the feature using the same scaler used for training
                if feature in self.feature_columns:
                    # Get the index of the feature in the original feature columns
                    original_idx = self.feature_columns.index(feature)
                    # Scale the feature
                    feature_vector[0, i] = (audio_profile[feature] - self.scaler.mean_[original_idx]) / self.scaler.scale_[original_idx]
                else:
                    # For engineered features, just use the raw value
                    feature_vector[0, i] = audio_profile[feature]

        # Generate embedding for the profile
        profile_embedding = self.encoder_model.predict(feature_vector)

        # Calculate similarity to all tracks
        embedding_cols = [col for col in embedding_df.columns if col.startswith('embedding_')]
        all_embeddings = embedding_df[embedding_cols].values
        similarities = cosine_similarity(profile_embedding, all_embeddings)[0]

        # Add similarities to a copy of the dataframe
        results_df = embedding_df.copy()
        results_df['similarity'] = similarities

        # Sort by similarity
        recommendations = results_df.sort_values('similarity', ascending=False).head(n)

        return recommendations

    def save_model(self, model_path='content_based_model'):
        """Save the trained encoder model and preprocessing components"""
        if self.encoder_model is None:
            print("No encoder model to save. Train the autoencoder first.")
            return False

        # Create directory if it doesn't exist
        if not os.path.exists(model_path):
            os.makedirs(model_path)

        # Save encoder model
        self.encoder_model.save(os.path.join(model_path, 'encoder_model.keras'))

        # Save autoencoder model if available
        if hasattr(self, 'autoencoder'):
            self.autoencoder.save(os.path.join(model_path, 'autoencoder_model.keras'))

        # Save components
        import pickle

        # Save scaler
        with open(os.path.join(model_path, 'feature_scaler.pkl'), 'wb') as f:
            pickle.dump(self.scaler, f)

        # Save feature columns
        with open(os.path.join(model_path, 'feature_columns.txt'), 'w') as f:
            f.write('\n'.join(self.feature_columns))

        # Save all feature columns (including engineered features)
        with open(os.path.join(model_path, 'all_feature_columns.txt'), 'w') as f:
            f.write('\n'.join(self.all_feature_columns))

        print(f"Model and components saved to {model_path}")
        return True

    def load_model(self, model_path='content_based_model'):
        """Load a saved encoder model and preprocessing components"""
        if not os.path.exists(model_path):
            print(f"Model path {model_path} does not exist.")
            return False

        # Load encoder model
        encoder_path = os.path.join(model_path, 'encoder_model.keras')
        if os.path.exists(encoder_path):
            self.encoder_model = tf.keras.models.load_model(encoder_path)
        else:
            print(f"Encoder model not found at {encoder_path}")
            return False

        # Load autoencoder model if available
        autoencoder_path = os.path.join(model_path, 'autoencoder_model.keras')
        if os.path.exists(autoencoder_path):
            self.autoencoder = tf.keras.models.load_model(autoencoder_path)

        # Load components
        import pickle

        # Load scaler
        scaler_path = os.path.join(model_path, 'feature_scaler.pkl')
        if os.path.exists(scaler_path):
            with open(scaler_path, 'rb') as f:
                self.scaler = pickle.load(f)

        # Load feature columns
        feature_columns_path = os.path.join(model_path, 'feature_columns.txt')
        if os.path.exists(feature_columns_path):
            with open(feature_columns_path, 'r') as f:
                self.feature_columns = f.read().splitlines()

        # Load all feature columns
        all_feature_columns_path = os.path.join(model_path, 'all_feature_columns.txt')
        if os.path.exists(all_feature_columns_path):
            with open(all_feature_columns_path, 'r') as f:
                self.all_feature_columns = f.read().splitlines()

        print(f"Model and components loaded from {model_path}")
        return True


def main():
    # Initialize the recommender
    sample_size = 150000  # Adjust based on available memory
    recommender = ContentBasedMusicRecommender('spotify_data.csv', chunk_size=50000, sample_size=sample_size)

    # Process a sample of the data
    sample_data_path = recommender.process_sample_data()

    # Prepare data for training
    X_train, X_val, df = recommender.prepare_data_for_training(sample_data_path)

    # Build autoencoder
    input_dim = X_train.shape[1]
    autoencoder, encoder = recommender.build_autoencoder(input_dim, encoding_dim=32)

    # Train autoencoder
    recommender.train_autoencoder(X_train, X_val, epochs=30, batch_size=256)

    # Create embeddings for all tracks
    embedding_df = recommender.create_embeddings(sample_data_path)

    # Save model
    recommender.save_model()

    # Example: Find similar tracks to a random track
    if embedding_df is not None and len(embedding_df) > 0:
        # Get a random track ID
        random_track_id = embedding_df['track_id'].sample(1).iloc[0]

        # Find track details
        track_details = embedding_df[embedding_df['track_id'] == random_track_id]
        track_name = track_details['track_name'].iloc[0] if 'track_name' in track_details.columns else "Unknown"
        artist_name = track_details['artist_name'].iloc[0] if 'artist_name' in track_details.columns else "Unknown"

        print(f"\nFinding similar tracks to: {artist_name} - {track_name} (ID: {random_track_id})")

        # Find similar tracks
        similar_tracks = recommender.find_similar_tracks(random_track_id, embedding_df, n=10)

        if similar_tracks is not None:
            print("\nTop 10 Similar Tracks:")
            for i, (_, track) in enumerate(similar_tracks.iterrows()):
                artist = track.get('artist_name', 'Unknown')
                name = track.get('track_name', f"Track {track['track_id']}")
                similarity = track['similarity']
                print(f"{i+1}. {artist} - {name} (Similarity: {similarity:.4f})")


if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.regularizers import l2
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import mean_squared_error
import tensorflow as tf
import gc
import os
from tqdm import tqdm

class ContentBasedMusicRecommender:
    def __init__(self, spotify_data_path='spotify_data.csv', chunk_size=50000, sample_size=200000):
        """Initialize a content-based music recommender that doesn't use ratings"""
        self.spotify_data_path = spotify_data_path
        self.chunk_size = chunk_size
        self.sample_size = sample_size
        self.encoder_model = None
        self.feature_columns = [
            'danceability', 'energy', 'loudness', 'speechiness',
            'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo'
        ]
        self.scaler = StandardScaler()
        self.metrics = {
            'train_mse': [],
            'val_mse': [],
            'train_rmse': [],
            'val_rmse': [],
            'reconstruction_mse': None,
            'reconstruction_rmse': None
        }

        # Set numeric precision to float32
        tf.keras.backend.set_floatx('float32')

        # Configure TensorFlow for better numerical stability
        tf.config.optimizer.set_jit(False)  # Disable XLA to avoid NaN issues

        # Enable memory growth for GPU if available
        gpus = tf.config.experimental.list_physical_devices('GPU')
        if gpus:
            try:
                for gpu in gpus:
                    tf.config.experimental.set_memory_growth(gpu, True)
                print(f"Memory growth enabled on {len(gpus)} GPU(s)")
            except RuntimeError as e:
                print(f"Error setting memory growth: {e}")

    def process_sample_data(self, output_path='content_sample_data.csv'):
        """Process a sample of the data for content-based similarity"""
        print(f"Processing {self.sample_size} sample rows...")

        # Read and process data in chunks until we have enough samples
        processed_rows = 0
        sample_data = []

        for chunk in tqdm(pd.read_csv(self.spotify_data_path, chunksize=self.chunk_size)):
            # Process chunk
            processed_chunk = self.process_chunk(chunk)

            # Determine how many rows to take from this chunk
            rows_to_take = min(len(processed_chunk), self.sample_size - processed_rows)
            if rows_to_take <= 0:
                break

            # Add to sample data
            sample_data.append(processed_chunk.iloc[:rows_to_take])
            processed_rows += rows_to_take

            # Check if we have enough data
            if processed_rows >= self.sample_size:
                break

            # Clean up memory
            del processed_chunk
            gc.collect()

        # Combine all samples
        combined_sample = pd.concat(sample_data)

        # Add additional engineered features
        combined_sample = self.add_engineered_features(combined_sample)

        # Save to CSV
        combined_sample.to_csv(output_path, index=False)

        print(f"Sample processing complete. {len(combined_sample)} rows processed and saved to {output_path}")

        # Clean up memory
        del sample_data, combined_sample
        gc.collect()

        return output_path

    def process_chunk(self, chunk):
        """Process a single chunk of data with improved handling"""
        # Create a copy to avoid modifying the original
        processed = chunk.copy()

        # Keep only necessary columns: track_id, artist_name, track_name, and audio features
        keep_columns = ['track_id', 'artist_name', 'track_name'] + self.feature_columns
        processed = processed[[col for col in keep_columns if col in processed.columns]]

        # Handle missing values in audio features
        for feature in self.feature_columns:
            if feature in processed.columns:
                # Use median for missing values
                median_value = processed[feature].median()
                processed[feature] = processed[feature].fillna(median_value)

                # Clip extreme values to prevent instability
                q1 = processed[feature].quantile(0.01)
                q99 = processed[feature].quantile(0.99)
                processed[feature] = processed[feature].clip(q1, q99)

        # Scale audio features if present
        if all(feature in processed.columns for feature in self.feature_columns):
            # Fit scaler if not already fit
            if not hasattr(self.scaler, 'mean_'):
                self.scaler.fit(processed[self.feature_columns])

            processed[self.feature_columns] = self.scaler.transform(processed[self.feature_columns])

        return processed

    def add_engineered_features(self, df):
        """Add engineered features to improve model performance"""
        # Create interaction features between audio characteristics
        df['energy_loudness'] = df['energy'] * df['loudness']
        df['dance_valence'] = df['danceability'] * df['valence']
        df['acoustic_instrumental'] = df['acousticness'] * df['instrumentalness']

        # Create feature to represent electronic vs. acoustic spectrum
        df['electronic_score'] = df['energy'] - df['acousticness']

        # Create feature to represent vocal vs. instrumental spectrum
        df['vocal_score'] = df['speechiness'] - df['instrumentalness']

        # Create mood feature (combination of valence and energy)
        df['mood_score'] = 0.5 * df['valence'] + 0.5 * df['energy']

        # Create danceability intensity feature
        df['dance_intensity'] = df['danceability'] * df['energy']

        # Add genre information if available
        if 'genre' in df.columns:
            # One-hot encode the genre (simplified here)
            genre_dummies = pd.get_dummies(df['genre'], prefix='genre')
            df = pd.concat([df, genre_dummies], axis=1)

        return df

    def prepare_data_for_training(self, data_path):
        """Prepare data for training the autoencoder"""
        print("Preparing data for autoencoder training...")

        # Load sample data
        df = pd.read_csv(data_path)

        # Extract feature columns (original and engineered)
        feature_cols = [col for col in df.columns if col not in ['track_id', 'artist_name', 'track_name']]

        # Create a copy of the features as both input and target for autoencoder
        features = df[feature_cols].values

        # Split into train and validation sets
        X_train, X_val = train_test_split(features, test_size=0.2, random_state=42)

        print(f"Training data shape: {X_train.shape}")
        print(f"Validation data shape: {X_val.shape}")
        print(f"Number of features: {len(feature_cols)}")

        # Save feature column names for later reference
        self.all_feature_columns = feature_cols

        return X_train, X_val, df

    def build_autoencoder(self, input_dim, encoding_dim=32):
        """Build an autoencoder model to create a compact representation of audio features"""
        # Input layer
        input_layer = Input(shape=(input_dim,), name='input')

        # Encoder
        encoded = Dense(64, activation='relu', kernel_regularizer=l2(1e-4), name='encode1')(input_layer)
        encoded = BatchNormalization(name='encode1_bn')(encoded)
        encoded = Dropout(0.3, name='encode1_dropout')(encoded)

        encoded = Dense(encoding_dim, activation='relu', kernel_regularizer=l2(1e-4), name='bottleneck')(encoded)

        # Decoder
        decoded = Dense(64, activation='relu', kernel_regularizer=l2(1e-4), name='decode1')(encoded)
        decoded = BatchNormalization(name='decode1_bn')(decoded)
        decoded = Dropout(0.3, name='decode1_dropout')(decoded)

        decoded = Dense(input_dim, activation='linear', name='output')(decoded)

        # Full autoencoder model
        autoencoder = Model(inputs=input_layer, outputs=decoded)

        # Encoder model for extracting embeddings
        encoder = Model(inputs=input_layer, outputs=encoded)

        # Compile the model
        autoencoder.compile(
            optimizer=Adam(learning_rate=0.001),
            loss='mse'
        )

        self.autoencoder = autoencoder
        self.encoder_model = encoder

        print("Autoencoder built successfully:")
        autoencoder.summary()

        return autoencoder, encoder

    def train_autoencoder(self, X_train, X_val, epochs=30, batch_size=256):
        """Train the autoencoder model with detailed MSE and RMSE metrics"""
        if self.autoencoder is None:
            print("No autoencoder model created. Call build_autoencoder() first.")
            return None

        # Create callbacks
        callbacks = [
            EarlyStopping(
                monitor='val_loss',
                patience=5,
                restore_best_weights=True,
                verbose=1
            ),
            ReduceLROnPlateau(
                monitor='val_loss',
                factor=0.5,
                patience=3,
                min_lr=0.0001,
                verbose=1
            )
        ]

        # Create a custom callback to calculate RMSE at each epoch
        class MetricsCallback(tf.keras.callbacks.Callback):
            def __init__(self, X_train, X_val, metrics_dict):
                self.X_train = X_train
                self.X_val = X_val
                self.metrics_dict = metrics_dict

            def on_epoch_end(self, epoch, logs=None):
                # Predict on training data
                train_pred = self.model.predict(self.X_train, verbose=0)
                train_mse = mean_squared_error(self.X_train, train_pred)
                train_rmse = np.sqrt(train_mse)

                # Predict on validation data
                val_pred = self.model.predict(self.X_val, verbose=0)
                val_mse = mean_squared_error(self.X_val, val_pred)
                val_rmse = np.sqrt(val_mse)

                # Store metrics
                self.metrics_dict['train_mse'].append(train_mse)
                self.metrics_dict['val_mse'].append(val_mse)
                self.metrics_dict['train_rmse'].append(train_rmse)
                self.metrics_dict['val_rmse'].append(val_rmse)

                # Print metrics
                print(f"\nEpoch {epoch+1} - Train MSE: {train_mse:.6f}, Train RMSE: {train_rmse:.6f}")
                print(f"Epoch {epoch+1} - Val MSE: {val_mse:.6f}, Val RMSE: {val_rmse:.6f}")

        # Add the custom callback
        metrics_callback = MetricsCallback(X_train, X_val, self.metrics)
        callbacks.append(metrics_callback)

        # Train the model
        history = self.autoencoder.fit(
            X_train, X_train,  # Autoencoder learns to reproduce its input
            batch_size=batch_size,
            epochs=epochs,
            validation_data=(X_val, X_val),
            callbacks=callbacks,
            verbose=1
        )

        # Final evaluation on the entire dataset
        X_combined = np.vstack((X_train, X_val))
        reconstructed = self.autoencoder.predict(X_combined)

        # Calculate final MSE and RMSE
        final_mse = mean_squared_error(X_combined, reconstructed)
        final_rmse = np.sqrt(final_mse)

        self.metrics['reconstruction_mse'] = final_mse
        self.metrics['reconstruction_rmse'] = final_rmse

        print(f"\nFinal Reconstruction MSE: {final_mse:.6f}")
        print(f"Final Reconstruction RMSE: {final_rmse:.6f}")

        # Plot training history
        plt.figure(figsize=(15, 10))

        # Plot MSE
        plt.subplot(2, 1, 1)
        plt.plot(self.metrics['train_mse'], label='Train MSE')
        plt.plot(self.metrics['val_mse'], label='Validation MSE')
        plt.title('Autoencoder MSE')
        plt.ylabel('Mean Squared Error')
        plt.xlabel('Epoch')
        plt.legend()
        plt.grid(True)

        # Plot RMSE
        plt.subplot(2, 1, 2)
        plt.plot(self.metrics['train_rmse'], label='Train RMSE')
        plt.plot(self.metrics['val_rmse'], label='Validation RMSE')
        plt.title('Autoencoder RMSE')
        plt.ylabel('Root Mean Squared Error')
        plt.xlabel('Epoch')
        plt.legend()
        plt.grid(True)

        plt.tight_layout()
        plt.savefig('autoencoder_metrics.png')
        plt.show()

        return history, self.metrics

    def create_embeddings(self, data_path):
        """Create embeddings for all tracks using the trained encoder"""
        if self.encoder_model is None:
            print("No encoder model available. Train the autoencoder first.")
            return None

        # Load data
        df = pd.read_csv(data_path)

        # Extract features
        features = df[self.all_feature_columns].values

        # Generate embeddings
        embeddings = self.encoder_model.predict(features)

        # Add embeddings to dataframe
        embedding_df = pd.DataFrame(
            embeddings,
            columns=[f'embedding_{i}' for i in range(embeddings.shape[1])]
        )

        # Add track information
        embedding_df['track_id'] = df['track_id'].values
        if 'artist_name' in df.columns:
            embedding_df['artist_name'] = df['artist_name'].values
        if 'track_name' in df.columns:
            embedding_df['track_name'] = df['track_name'].values

        print(f"Created embeddings with shape: {embeddings.shape}")

        return embedding_df

    def find_similar_tracks(self, track_id, embedding_df, n=10):
        """Find similar tracks based on embedding similarity"""
        # Locate the reference track
        if track_id not in embedding_df['track_id'].values:
            print(f"Track ID {track_id} not found in embeddings.")
            return None

        # Get the track's embedding
        track_embedding = embedding_df[embedding_df['track_id'] == track_id].iloc[0]
        embedding_cols = [col for col in embedding_df.columns if col.startswith('embedding_')]
        track_vector = track_embedding[embedding_cols].values.reshape(1, -1)

        # Calculate similarity to all other tracks
        all_embeddings = embedding_df[embedding_cols].values
        similarities = cosine_similarity(track_vector, all_embeddings)[0]

        # Add similarities to a copy of the dataframe
        results_df = embedding_df.copy()
        results_df['similarity'] = similarities

        # Sort by similarity (excluding the reference track)
        similar_tracks = results_df[results_df['track_id'] != track_id].sort_values(
            'similarity', ascending=False
        ).head(n)

        return similar_tracks

    def evaluate_reconstruction_quality(self, X_sample, n_examples=5):
        """Evaluate reconstruction quality with MSE and RMSE metrics for specific examples"""
        if self.autoencoder is None:
            print("No autoencoder model available. Train the model first.")
            return None

        # Get a sample of data points
        if len(X_sample) > n_examples:
            indices = np.random.choice(len(X_sample), n_examples, replace=False)
            X_examples = X_sample[indices]
        else:
            X_examples = X_sample

        # Get reconstructions
        reconstructions = self.autoencoder.predict(X_examples)

        # Calculate metrics for each example
        example_metrics = []
        for i in range(len(X_examples)):
            example_mse = mean_squared_error(X_examples[i], reconstructions[i])
            example_rmse = np.sqrt(example_mse)
            example_metrics.append({
                'example_id': i,
                'mse': example_mse,
                'rmse': example_rmse
            })

        # Print metrics
        print("\nReconstruction Quality Metrics for Sample Examples:")
        print("=" * 50)
        for metric in example_metrics:
            print(f"Example {metric['example_id']+1}:")
            print(f"  MSE: {metric['mse']:.6f}")
            print(f"  RMSE: {metric['rmse']:.6f}")

        # Calculate average metrics
        avg_mse = np.mean([m['mse'] for m in example_metrics])
        avg_rmse = np.mean([m['rmse'] for m in example_metrics])

        print("\nAverage Metrics:")
        print(f"  Avg MSE: {avg_mse:.6f}")
        print(f"  Avg RMSE: {avg_rmse:.6f}")

        return example_metrics

    def save_model_and_metrics(self, model_path='content_based_model'):
        """Save the trained encoder model, preprocessing components, and metrics"""
        if self.encoder_model is None:
            print("No encoder model to save. Train the autoencoder first.")
            return False

        # Create directory if it doesn't exist
        if not os.path.exists(model_path):
            os.makedirs(model_path)

        # Save encoder model
        self.encoder_model.save(os.path.join(model_path, 'encoder_model.keras'))

        # Save autoencoder model if available
        if hasattr(self, 'autoencoder'):
            self.autoencoder.save(os.path.join(model_path, 'autoencoder_model.keras'))

        # Save components
        import pickle

        # Save scaler
        with open(os.path.join(model_path, 'feature_scaler.pkl'), 'wb') as f:
            pickle.dump(self.scaler, f)

        # Save feature columns
        with open(os.path.join(model_path, 'feature_columns.txt'), 'w') as f:
            f.write('\n'.join(self.feature_columns))

        # Save all feature columns (including engineered features)
        with open(os.path.join(model_path, 'all_feature_columns.txt'), 'w') as f:
            f.write('\n'.join(self.all_feature_columns))

        # Save metrics
        with open(os.path.join(model_path, 'metrics.pkl'), 'wb') as f:
            pickle.dump(self.metrics, f)

        # Save a metrics report
        with open(os.path.join(model_path, 'metrics_report.txt'), 'w') as f:
            f.write("AUTOENCODER PERFORMANCE METRICS\n")
            f.write("==============================\n\n")

            f.write("Final Reconstruction Metrics:\n")
            f.write(f"MSE: {self.metrics['reconstruction_mse']:.6f}\n")
            f.write(f"RMSE: {self.metrics['reconstruction_rmse']:.6f}\n\n")

            f.write("Training Progress:\n")
            f.write("Epoch, Train MSE, Train RMSE, Val MSE, Val RMSE\n")
            for i in range(len(self.metrics['train_mse'])):
                f.write(f"{i+1}, {self.metrics['train_mse'][i]:.6f}, {self.metrics['train_rmse'][i]:.6f}, ")
                f.write(f"{self.metrics['val_mse'][i]:.6f}, {self.metrics['val_rmse'][i]:.6f}\n")

        print(f"Model, components, and metrics saved to {model_path}")
        return True


def main():
    # Initialize the recommender
    sample_size = 150000  # Adjust based on available memory
    recommender = ContentBasedMusicRecommender('spotify_data.csv', chunk_size=50000, sample_size=sample_size)

    # Process a sample of the data
    sample_data_path = recommender.process_sample_data()

    # Prepare data for training
    X_train, X_val, df = recommender.prepare_data_for_training(sample_data_path)

    # Build autoencoder
    input_dim = X_train.shape[1]
    autoencoder, encoder = recommender.build_autoencoder(input_dim, encoding_dim=32)

    # Train autoencoder with detailed metrics
    history, metrics = recommender.train_autoencoder(X_train, X_val, epochs=30, batch_size=256)

    # Create embeddings for all tracks
    embedding_df = recommender.create_embeddings(sample_data_path)

    # Evaluate reconstruction quality on a sample of examples
    sample_size_to_evaluate = min(100, len(X_val))
    X_sample = X_val[:sample_size_to_evaluate]
    recommender.evaluate_reconstruction_quality(X_sample, n_examples=5)

    # Save model and metrics
    recommender.save_model_and_metrics()

    # Print final metrics summary
    print("\nFINAL METRICS SUMMARY")
    print("====================")
    print(f"Final Reconstruction MSE: {metrics['reconstruction_mse']:.6f}")
    print(f"Final Reconstruction RMSE: {metrics['reconstruction_rmse']:.6f}")

    # Example: Find similar tracks to a random track
    if embedding_df is not None and len(embedding_df) > 0:
        # Get a random track ID
        random_track_id = embedding_df['track_id'].sample(1).iloc[0]

        # Find track details
        track_details = embedding_df[embedding_df['track_id'] == random_track_id]
        track_name = track_details['track_name'].iloc[0] if 'track_name' in track_details.columns else "Unknown"
        artist_name = track_details['artist_name'].iloc[0] if 'artist_name' in track_details.columns else "Unknown"

        print(f"\nFinding similar tracks to: {artist_name} - {track_name} (ID: {random_track_id})")

        # Find similar tracks
        similar_tracks = recommender.find_similar_tracks(random_track_id, embedding_df, n=10)

        if similar_tracks is not None:
            print("\nTop 10 Similar Tracks:")
            for i, (_, track) in enumerate(similar_tracks.iterrows()):
                artist = track.get('artist_name', 'Unknown')
                name = track.get('track_name', f"Track {track['track_id']}")
                similarity = track['similarity']
                print(f"{i+1}. {artist} - {name} (Similarity: {similarity:.4f})")


if __name__ == "__main__":
    main()
```
